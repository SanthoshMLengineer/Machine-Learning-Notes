**<h1><center>ANOVA TEST</center></h1>**

Analysis of Variance, or Anova, is a statistical test that determines how changes in a quantitative variable are influenced by values of a categorical independent variable. This test checks whether their is difference between means at every category of independent variable.

**<u>One-way Anova test</u>**

This test is used when we have one independent categorical variable and one dependent quantitative variable.This test determines whether dependent quantitative variable changes over categories of categorical independent variable.

<u>null hypothesis (H0)</u> : Their is no difference in means.

<u>alternate hypothesis (Ha)</u> : Their is atleast one group differs from others.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
subject =   ["Maths","Science"]

# Generate random data
subject = np.random.choice(a= subject,
                              p = [0.40, 0.60],
                              size=500)


marks = stats.poisson.rvs(loc=18,
                          mu=30,
                          size=500)

# Group age data by race
data = pd.DataFrame({"subject":subject,"marks":marks})

data
```

```python
groups = data.groupby("subject").groups
```

```python
groups["Maths"]
```

```python
# Etract individual groups
maths = data.loc[groups["Maths"], ['marks']]
science = data.loc[groups["Science"],['marks']]
```

```python
science
```

```python
# Perform the ANOVA
F_statistic, p_value = stats.f_oneway(maths, science)

print("F-statistic:", F_statistic)
print("p-value:", p_value)

if p_value < 0.05:
    print("The difference in means between the groups is statistically significant.")
else:
    print("There is no statistically significant difference in means between the groups.")
```
