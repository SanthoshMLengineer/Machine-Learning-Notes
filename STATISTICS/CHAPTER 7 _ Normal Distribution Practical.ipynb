**<center><h1>Normal Distribution</h1></center>**

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import numpy as np
from scipy.stats import kurtosis
from scipy.stats import skew

mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 1000)

sns.distplot(s)
```

```python
print( 'excess kurtosis of normal distribution (should be 0): {}'.format( kurtosis(s) ))
print( 'skewness of normal distribution (should be 0): {}'.format( skew(s) ))

# skewness between -0.5 and 0.5, we consider the distribution to be approximately symmetric
# Kurtosis value will be close to Zero for symmetric type of distribution.
```

```python
left_skewed = [10,20,15,30,23,35,40,32,50,50,50,50.5,60,60,60,60,60,70,70,70,80,80,80,90]
sns.distplot(left_skewed)
```

```python
print( 'excess kurtosis of normal distribution (should be 0): {}'.format( kurtosis(left_skewed) ))
print( 'skewness of normal distribution (should be 0): {}'.format( skew(left_skewed) ))
```

```python
left_skewed = np.tanh(left_skewed)

sns.distplot(left_skewed)
```

```python
right_skewed = [565, 116, 69, 78, 319, 501, 633, 655, 848, 1226, 1633, 2459, 3375, 4669, 6152, 7436, 9526, 12619, 12455, 7113, 2104, 241]
sns.distplot(right_skewed)
```

```python
print( 'excess kurtosis of normal distribution (should be 0): {}'.format( kurtosis(right_skewed) ))
print( 'skewness of normal distribution (should be 0): {}'.format( skew(right_skewed) ))
```

```python
right_skewed = np.log(right_skewed)

sns.distplot(right_skewed)
```

```python
print( 'excess kurtosis of normal distribution (should be 0): {}'.format( kurtosis(right_skewed) ))
print( 'skewness of normal distribution (should be 0): {}'.format( skew(right_skewed) ))
```
