**<h3>Random Variable</h3>**

Random variable is a variable which stores outcome of an experiment.

`Example : Let's take an experiment of flipping a coin, let's use X variable
X = 0 if the flip of the coin is a head.
X = 1 if the flip of the coin is a tail.
Here we used X variable to represent the outcome of an experiment these
type of variables called Random Variables.`

Their are two types of random variable:

1. Discrete Random Variable
2. Continuous Random Variable
    
**<u>1. Discrete Random Variable</u>**

The random variables is a variable which stores distinct values.

`For example: Consider a random variable X representing coin flip. All possible outcome of this experiments can be listed as {0,1}. In this case, the possible outcomes of the experiment are countable. Hence this is a discrete random variable.`
    
**<u>2. Continuous Random Variable</u>**

Continuous random variables are those variables which will take any value between a range.

`For example: Consider an experiment where the weight of a university student is chosen at random .`

**Normal Distribution**

Normal Distribution is also known as gaussian distribution. It is a type of continuous probability Distribution in which the majority of the data group together around the mean.
When we plot graph for random variable looks like bell shaped curve.

<center><img src="https://miro.medium.com/max/558/1*i2Nf81cjZHRhT2v5chUIYw.jpeg"></center>


- Above graph tells all data points clustered in center which is mean.
- In normal distribution Mean, Median, and Mode are all the same.

**Emperical Rule**
Emperical rule says that
- 68.27% of data lies within 1 standard deviation of the mean
- 95.45% of data lies within 2 standard deviations of the mean
- 99.73% of data lies within 3 standard deviations of the mean
    
**<h3>Skewed Distribution</h3>**

Normal Distribution is symmetric, which means looks like bell shaped curve where its its tails on one side are the mirror image of the other side. But with most datasets, this is not the case. Generally, data points cluster on one side more than the other. These distributions are known as Skewed Distributions.


**1. Left Skewed Distribution**

When data points cluster on the right side of the distribution, then the tail would be longer on the left side. This is the property of Left Skewed Distribution. This skewness is called Negatively skewed distribution where tail is longer in left side.

<center><img src="https://www.statisticshowto.com/wp-content/uploads/2015/07/skewed-left.png"></center>
    
Negatively skewed distribution, we express the general relationship between the central tendency measured as `Mode > Median > Mean`
        
        
**2. Right Skewed Distribution**
When data points cluster on the left side of the distribution, then the tail would be longer on the right side. This is the property of Right Skewed Distribution. This skewness is called Positively skewed distribution where tail is longer in right side.

<center><img src="https://i1.wp.com/psychologyhub.co.uk/wp-content/uploads/2018/01/right-skewed-graph.jpg?w=367&h=336&ssl=1"></center>
    
Positively skewed distribution, we express the general relationship between the central tendency measured as: `Mode < Median < Mean`

**<h3><u>How to check normality?</u></h3>**

**1. Histogram**
A histogram shows how data are distributed across a continuous range.
In a histogram, each bar corresponds to the tabulated frequency at each interval or bin. In simple words, height represents the frequency for the respective bin (interval)

<center><img src="https://www150.statcan.gc.ca/edu/power-pouvoir/c-g/c-g05-7-1-eng.png" width = 350px></center>

**2. Skewness**
Skewness is also another measure to check for normality which tells us the amount and direction of the skewed data points.

Generally for the value of Skewness:

- If the value is less than -0.5, we consider the distribution to be negatively skewed or left-skewed.
-  Whereas if the value is greater than 0.5, we consider the distribution to be positively skewed or right-skewed.
- And finally, if the value is between -0.5 and 0.5, we consider the distribution to be approximately symmetric


**3. kurtosis**

Kurtosis refers to the degree of presence of outliers in the distribution. Kurtosis is a statistical measure, whether the data is heavy-tailed or light-tailed in a normal distribution.
  
Types of Kurtosis
- Leptokurtic or heavy-tailed distribution (kurtosis more than normal distribution , kurtosis > 3).
- Mesokurtic (kurtosis same as the normal distribution,  kurtosis is near to 3).
- Platykurtic or short-tailed distribution (kurtosis less than normal distribution, kurtosis < 3).
