```python
# Importing Required Packages
import pandas as pd

data = pd.read_csv("/content/drive/My Drive/Selling Project/\
Real Time Projects/Loan Default Classification/Data Files/train.csv")
```

```python
print("Columns in data set\n",data.columns)
print("{} rows  and {} columns in data set".format(data.shape[0],data.shape[1]))
```

```python
# describe function gives descriptive stats summary of input data set.
data.describe()
```

```python
# If we want to find out mean,median and mode we can do with pandas.
print("mean of ApplicantIncome column is {} ".format(data['ApplicantIncome'].mean()))
```

```python
data['ApplicantIncome'].plot(kind = 'box')
```

```python
# Above box plot tells us their is lots of outliers so mean is not correct measure of central tendency, so we use median.
print("median of ApplicantIncome column is {} ".format(data['ApplicantIncome'].median()))
```

#### Percentiles and Quartiles

```python
import numpy as np
print("25 percentile of ApplicantIncome column {}".format(np.percentile(data['ApplicantIncome'],25)))
print("45 percentile of ApplicantIncome column {}".format(np.percentile(data['ApplicantIncome'],45)))
```

```python
print("25,50,75 quartiles for ApplicantIncome column\n\n",data['ApplicantIncome'].quantile([0.25,0.5,0.75]))
print("\n\n InterQuartile range is {}".format(data['ApplicantIncome'].quantile(0.75) - data['ApplicantIncome'].quantile(0.25)))
```
