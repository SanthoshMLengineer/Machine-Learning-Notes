**<h1><center>Descriptive Statistics</center></h1>**

Descriptive Statistics is all about summarizing the data like mean, median mode etc. It gives some basic information about data whichc helpds further data analysis process.

Two types of Descriptive statistics:
  - Measure of Central Tendency
  - Measure of Spread / Dispersion 


**<h3>1. Measure of Central Tendency</h3>**

A basic and initial step is to explore your data is to get typical value which gives where most of the data is located. 

Methods to describe typical value for the feature :

- Mean or Average
- Trimmed Mean
- Median
- Mode

**Mean or Average**

Mean is calculated by summing all the values divided by the total number of values, where total number of values is represented as `N (or n)`, `N` for population and `n` for sample.
-  Example : 

  ```
  Mean for 20, 30, 40 is 
          
          = (20+30+40)/ total number of values

          = (90)/3

          = 30
  ```


**Trimmed Mean or Truncated mean**

- Trimmed mean is same as mean, but it trims the outliers. You can specify how much percentage data you want to remove. 
- For Example with a 5% trimmed mean, the lowest 5% and highest 5% of the data are excluded. The mean is calculated from the remaining 90% of data points.

  ```
  Find the trimmed 20% mean for the following test scores: 
          60, 81, 83, 91, 99.

  Step 1: Trim the top and bottom 20% from the data. 
          That leaves us with the middle three values:60, 81, 83, 91, 99.

  Step 2: Find the mean with the remaining values. 
          The mean is (81 + 83 + 91) / 3 ) = 85.

  ```

**Median**

- Median is nothing but getting middle number of sorted data. N or n is used to represent total number of values. 
```
If N(or n) is ODD then median is a value at position (N + 1)/2.
If N(or n) is even then we need to calculte value 
        First_value is (N)/2 and value 
        Second_value is (N + 1)/2 then 
        median is (First_value + Second_value )/2. 
```


**Mode**

- Mode is a value which having highest frequency in a list of values. 

  ```
  Example : 2, 6, 7, 8, 8
  Mode value is 8 which having two occurrences.
  ```

**<h3>2. Measure of Spread/Dispersion/Variability</h3>**

Dispersion or variability is used to find out whether the data values are tightly clustered or spread out.

Different Ways to find  out Dispersion

- Varience
- Standard Deviation
- Mean Deviation / Mean Absolute Deviation
- Range
- Percentile
- Quartiles


**Varience**

Varience is square average distance between each value and mean.

```
Where S^2 is sample Varience,

      x is a value of the dataset, 

      x̅ is mean value

      N is total number of values in dataset

Example

    Values = 600,470,170,430,300 and Mean = 394 (Considering as population)
    Varience = (600-394)^2 + (470-394)^2 + (170-394)^2 + (430-394)^2 + (300-394)^2 /5
            = 21704
      
```

**Standard Deviation**

Standard Deviation is average distance between each value and mean. Low Standard deviation indicates that data points are close to mean of the data, while a high Standard Deviation indicated that data points are spread out over mean.

Standard Deviation is square root of varience

  ```

  Example
      Varience = 21704
      Standard Deviation  = Sq(Varience) = 147
  ```

**Mean Deviation / Mean Absolute Deviation**

It is an average of differences between each value and mean.

  ```
  Example
      
      Values = 600,470,170,430,300 and Mean = 394 (Considering as population)
      Mean Absolute Deviation = 
              
              (600-394) + (470-394) + (170-394) + (430-394) + (300-394) /5 = 
              
              206 + 76 + (−224) + 36 + (−94) / 5 = 0 /5 = 0 
  ```

**Range**

It is difference between lowest and highest values.

  ```
  Example : 2, 6, 7, 8, 9
  
      Range = 9 - 2 = 7

  ```

**Percentile**

Percentile is a way to represent position of a values in data set. To calculate percentile, values in data set should always be in ascending order.
  
<u>Quartiles</u>
    
  There are three quartile values. 
  - First quartile(Q1) value is at 25 percentile and it is median of upper half of the data. 
  - Second quartile(Q2) is 50 percentile ad it is median of the whole data.
  - Third Quartile (Q3) is median of lower half of the data and third quartile is 75 percentile. 

  ```
  Example : 12 , 24, 41, 51, 67, 67, 85, 99, 115

      Q1 = 41: is 25 percentile of the data.

      Q2 = 67: is 50 percentile of the whole data and that is median.
      
      Q3 = 85: is 75 percentile of the date.
      
      Interquartile range (IQR) = Q3 - Q1 = 85 - 41 = 44
  ```




**<h3>Outliers</h3>**

An outlier is any value that is very distant from the other values in a data set. In fact, a trimmed mean and median is widely used to avoid the influence of outliers.

**How to detect Outliers?**


Most commonly used method to detect outliers is visualization. We use boxplot to detect outliers.

<img src = "https://codingwithmax.com/wp-content/uploads/2019/11/box-plot-vertical-horizontal-1.png">


Various rules to detect outliers. Some of them are:
- Any value, which is beyond the range of Q1 - 1.5 x IQR to 1.5 x IQR + Q3 .
- Any value which out of range of 5th and 95th percentile can be considered as outlier.
- Data points, three or more standard deviation away from mean are considered outlier.

**How to Treat outliers?**

- Deleting Observation
- Imputing
- Transforming and binning values
- Treat Separately.


**Note**
- An outlier can affect the mean of a data set.
- Median and trimmed mean is best if feature has outliers.
