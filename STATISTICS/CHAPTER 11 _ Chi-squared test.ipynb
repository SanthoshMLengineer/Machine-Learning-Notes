**<center><h1>Chi-Squared Test</h1></center>**

This test helps to determine whether their is relation between two categorical variable or both are independent.

<u>Hypothesis</u>

- H0: the two samples are independent.
- H1: there is a dependency between the samples.


**<u>Steps to conduct chisquare test</u>**

1. <u>Get Expected values</u> : Expected values are values will be their if their is no association between two variables.

      Expected value of each cell = row total * column total / table total


2. <u>Get Observed values</u> : Observed values are values counted from the sample.


3. <u>Get Chi-square statistic</u> : It compares expected values and observed values. It calculates how far observed values from expected values.


4. <u>Get degree of freedom</u> : (rows - 1) * (column - 1) will gives the degree of freedom.

5. <u>Get test statistic</u> : This values will get from chi-square  distribution.


If chi-square-statistic value is greater than test-statistic-value then their is a relation between two variables else two variables are independent.

```python
import pandas as pd


data = {"Gender":["Male","Female","Male","Female","Female","Male","Male",
                  "Female","Female"],
        "Shopping":["No","Yes","Yes","Yes","Yes","Yes","No","No","No"]}

chi_data = pd.DataFrame(data)
```

```python
contingency_table = pd.crosstab(chi_data["Gender"],chi_data["Shopping"])

print('contingency_table :-\n',contingency_table)
```

1. Observed values

```python
#Observed Values
Observed_Values = contingency_table.values

print("Observed Values :-\n",Observed_Values)
```

2. Expected values

```python
import scipy.stats

chi_square_values = scipy.stats.chi2_contingency(contingency_table)


Expected_Values = chi_square_values[3]
print("Expected Values :-\n",Expected_Values, chi_square_values)
```

3. Degree of freedom

```python
no_of_rows = len(contingency_table.iloc[0:2,0])
no_of_columns = len(contingency_table.iloc[0,0:2])

ddof = (no_of_rows-1)*(no_of_columns-1)

print("Degree of Freedom:-",ddof)
```

4. chi-square statistic

```python
from scipy.stats import chi2

#Significance Level 5%
alpha = 0.05

chi_square = sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])

chi_square_statistic = chi_square[0]+chi_square[1]

print("chi-square statistic:-",chi_square_statistic, chi_square)
```

5. Test statistic

```python
critical_value=chi2.ppf(q=1-alpha,df=ddof)

print('critical_value:',critical_value)

```

```python
print('Degree of Freedom: ',ddof)
print('chi-square statistic:',chi_square_statistic)
print('critical_value:',critical_value)

if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```
