**<center><h1>Bayes Theorem</h1></center>**

Bayes theorem is a probabilistic algorithm which helps to find how likely an event can occur given that previous event already occured. Naive Bayes algorithm uses Bayes Theorem. 


<center><img src = "https://www.gstatic.com/education/formulas2/355397047/en/bayes__theorem.svg"></center>

A, B 	= 	events

P(A|B) 	= 	probability of A given B is true

P(B|A) 	= 	probability of B given A is true

P(A), P(B) 	= 	the independent probabilities of A and B


Types of Naive Bayes algorithm :

1. Bernoulli Naive Bayes : This algorithm is used for binary classification.

2. Multinomial Naive Bayes : This algorithm is used multiple classification.
