**<h1><center>STATISTICS</center></h1>**
 
Statistics is all about study, organizing, analysing and understanding underlying patterns of data. 

<u>**Terminologies in Statistics**</u>

1. Population contains every possible events in an experiment.
 
2. Parameter describes the characterstics of population such as population mean,median etc.
 
3. Sample is a subset of the population.
 
4. Statistics describes the characterstics of sample such as sample mean , meadian etc.

**<h3>1. Numerical or quantitative</h3>**
 
Quantitative data types are expressed in terms of numerical.
  `Example : Price , income etc.`

Their are two types of data in Numerical data type : 
1. **Continuous Data Type:**
  A continuous data type takes any values in certain range.
  `Example: Height, weight, length, temperature.`

2. **Discrete Data Type:**
  Discrete data type can take on certain individual values.
  `Example: Shirt size,Shoe size etc.`



**<h3>2. Categorical or qualitative</h3>**

Qualitative data type represents in the form text.
  `Example : marital status etc.`

Their are two types of data in Categorical data type.
1. **Nominal Data Type :**
The values which is not having specific order. 

  `Example: Names , Tv, fan etc.`

2. **Ordinal Data Type:**
The ordinal data in which the values or categories are ordered.

  `Example: Education Scoring Class (Fail, Pass, First Class, Second Class, Distinction) , Ageing(Young age , Middle Age , Old age) etc.`
            
3. **Binary Data Type:**
Binary data is an categorical data that takes on only one of two values.

  `Example: 0/1, yes/no, or true/false`




**<h3><u>Two types of statistics</u></h3>**

**1. Descriptive Statistics**

  In Descriptive Statistics it helps to summarize your data like mean, median, mean etc. It gives basic information about data helps to further proceed the data analysis.
      
**2. Inferential Statistics**
      
  It is all about making inferences(Conclusion) about popultions by taking sample from that population.
