**<h1><center>Correlation</center></h1>**

Correlation is a metric that tells strength of the relationship between two variable. It's always in between -1 to 1.

Correlation can be expressed as

- From 0.5 to 1 is strongly positive correlated.
- From 0 t0 0.5 moderately positive correlated.
- From -0.5 to -1 is stronly negatively correlated.
- From -0.5 to 0 moderately neagtive correlated.

```python
def sample_std(x):
    return np.std(x)


X = [16, 14, 19, 12, 19, 15]
Y = [ 2,  1, 11,  1, 14, 18]

def correlation(x,y):
    cov = covariance(x,y)
    return cov/(sample_std(x) * sample_std(y))

correlation(X,Y)
```
