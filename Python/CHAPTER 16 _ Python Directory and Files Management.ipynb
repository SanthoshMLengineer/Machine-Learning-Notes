**<h2>Python Directory</h2>**

A directory or folder is a collection of files and subdirectories. Python has os module that provides many useful function to work with directories.

- **Get Current Directory**

  - getcwd() method is used to get present working directory. 
    ```
    import os

    os.getcwd()
    ```

- **Changing Directory**

  - chdir() method helps to change current working directory into specified path. The path that we want to change that should be supplied as a string to this method. We can use both the forward-slash `/` or the backward-slash `\` to separate the path elements. 

    ```
    os.chdir('C:\\Python3')
    ```

- **List Directories and Files**

  - listdir() method is used retrieve all files and sub-directories inside a directory. This function accepts a directory path and returns a list of its subdirectories and files. If no path is supplied it returns the list of files and subdirectories from the current working directory. 

    ```
    os.listdir()
    ```

- **Making a New Directory**

  - mkdir() method is used to make new directory. Path of new directory should be given as parameter to this method. If the full path is not specified, the new directory is created in the current working directory.
    ```
    os.mkdir('test')
    ```

- **Renaming a Directory or a File**

  - The rename() method helps to rename a directory or a file. rename() method takes in two basic arguments: the old name as the first argument and the new name as the second argument.

    ```
    os.rename('test','new_one')` 
    # test is original name and new_one is modified name.
    ```

- **Removing Directory or File**

  - remove() method helps to delete a file. rmdir() helps to delete empty directory. 

    ```
    os.remove('old.txt')`# removes the file
    
    os.rmdir('new_one')` # removes directory,rmdir removes empty directory
    ```

  - rmtree() method helps to remove non-empty directory. rmtree() method inside the shutil module.

    ```
    import shutil
    
    shutil.rmtree('test')
    ```
