**<center><h1>Function</h1></center>**

Function is a group of statements that does specific task. Functions help break our program into smaller chunks. 

- Syntax of Function:
  ```
  def function_name(parameters):
      """docstring"""
      statement(s)
  ```




````python
def greet(name):
    """
    This function greets to
    the person passed in as
    a parameter
    """
    print("Hello, " + name + ". Good morning!")
````

Components of function:

- `def` keyword helps define function header. Function name should be unique and rules for naming a function as same as rules for naming identifiers. From above **greet** is the function name.

- Parameters (arguments) through which we pass values to a function. From above example **name** is a parameter. Parameters are option.

- A colon (:) to mark the end of the function header.

- Docstring is small documentation helps to understand what function does, and what are the input parameters and what it returns to calling function. One or more valid python statements that make up the function body. Statements must have the same indentation level (usually 4 spaces).
- Function can have return statments and it can optional.

**How to call a function in python?**

Once we have defined a function, we can call it from another function, program or even the Python prompt. To call a function we simply type the function name with appropriate parameters

````python
greet('Santhosh')
````

**<h3>Docstrings</h3>**

- The docstring is also known as the documentation string, it is the first string enclosed in triple quotes following the function header. It is used to describe what a function performs. This string is available to us as the doc attribute of the function.

````python
print(greet.__doc__)
````

**The return statement**

- The return statement returns values to the place where it was called.
- Syntax of return:
  ```
  return [expression_list]
  ```
- This statement can contain an expression that gets evaluated or value is returned. The function will return None object if there are no return statements. 

````python
print(greet("May"))
# Here, None is the returned value since greet() directly prints the name 
#and no return statement so it prints none
````

````python
def absolute_value(num):
    """
    This function returns the absolute
    value of the entered number
    """
    if num >= 0:
        return num
    else:
        return -num
        
print(absolute_value(2))
print(absolute_value(-4))
````

**Types of Functions**

  - User-defined functions - Functions defined by the users themselves.
