**<h2>Python Output function</h2>**

- Print() function is used to display the output data to output device like screen.
    `
      Syntax:
          print(str, object, sep)

          str : String wanted to displayed on the screeen.

          object : Any object that wanted to display.

          sep : seperate is used to seperate the objects. Default will be space.
    `

    `print('This is output function used to display output data.')`
    
- str.format() function is used format the output data to make it look attractive.

    ```
    x = 10
    y = 5
    print('The value of x is {} and y is {}'.format(x,y))
    ```
        
- Here, the curly braces {} are used as placeholders. We can specify the order in which they are printed by using numbers (tuple index).

    ```
    print('I love {0} and {1}'.format('python','Machine Learning '))
    print('I love {1} and {0}'.format('python','Machine Learning'))
    ```



````python
print('I love {0} and {1}'.format('python','Machine Learning'))
print('I love {1} and {0}'.format('python','Machine Learning'))
````

**<h2>Python Input function</h2>**

- Input() function is used to take input from the user to make code flexible.
- The syntax for input() is:

    ```
    input([prompt])
    ```
    
- where prompt is the string we wish to display on the screen. It is optional.

    ```
    num = input('Enter a number: ')
    Enter a number: 10
    ```

- Input to the python from keyword always is string data type , we convert it into required data type using type casting.

 ```
 int('10')
 float('10')
 10.0
 ```

````python
num = input('Enter a number: ')

print(type(num))
````

````python
num = int(input('Enter a number: '))

print(type(num))
````
