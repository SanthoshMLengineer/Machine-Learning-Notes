**<center><h1>Operators in Python</h1></center>**

- Operators are special symbols that are used for arithmetic or logical computation.

- Operators are :

  - Arithmetic Operators
  - Comparison operators
  - Logical operators
  - Bitwise operators
  - Assignment operators
  - Membership operators


**<h2>Arithmetic Operators in Python</h2>**

- These symbols are used for arithmetic operation like Addition, Substraction, Multiplication etc. The value that the operator operates on is called the operand.
  - " + "
    - Add two operands or unary plus	
        `x + y+ 2`

  - " - "	
    - Subtract right operand from the left or unary minus	
        `x - y- 2`

  - " * "	
    - Multiply two operands	
        `x * y`

  - " / "	
    - Divide left operand by the right one (always results into float)	
        `x / y`

  - " % "	
    - Modulus - remainder of the division of left operand by the right	
        `x % y`  (remainder of x/y)

  - " // "	
    - Floor division - division that results into whole number adjusted to the left in the number line	
        `x // y`

  - " ** "	
      - Exponent - left operand raised to the power of right	
        `x**y (x to the power y)`

````python
x = 15
y = 4

# Output: x + y = 19
print('x + y =',x+y)

# Output: x - y = 11
print('x - y =',x-y)

# Output: x * y = 60
print('x * y =',x*y)

# Output: x / y = 3.75
print('x / y =',x/y)

# Output: x // y = 3
print('x // y =',x//y)

# Output: x ** y = 50625
print('x ** y =',x**y)
````

**<h2>Comparison Operators in Python</h2>**


- Comparison operators are used to compare values. It returns either True or False according to the condition.

    - " > "	
        - Greater than - True if left operand is greater than the right	
        `x > y`
        
    - " < "	
        - Less than - True if left operand is less than the right	
        `x < y`
    - " == "	
        - Equal to - True if both operands are equal	
        `x == y`
    - " != "	
        - Not equal to - True if operands are not equal	
        `x != y`
    - " >= "	
        - Greater than or equal to - True if left operand is greater than or equal to the right	
        `x >= y`
    - " <= "	
        - Less than or equal to - True if left operand is less than or equal to the right	
        `x <= y`

````python
x = 10
y = 12

# Output: x > y is False
print('x > y is',x>y)

# Output: x < y is True
print('x < y is',x<y)

# Output: x == y is False
print('x == y is',x==y)

# Output: x != y is True
print('x != y is',x!=y)

# Output: x >= y is False
print('x >= y is',x>=y)

# Output: x <= y is True
print('x <= y is',x<=y)
````

**<h2>Logical Operators in Python</h2>**

- " and "
    - True if both the operands are true	
    `x and y`
    
    
- " or "
    - True if either of the operands is true	
    `x or y`
    
    
- " not "
    - True if operand is false (complements the operand)	
    `not x`

````python
x = True
y = False

print('x and y is',x and y)

print('x or y is',x or y)

print('not x is',not x)
````

**<h2>Bitwise Operators in Python</h2>**

- Bitwise operators act on binary digits. They operate bit by bit, hence name  Bitwise Operator.
- Let x = 10 (1010 in binary) and y = 4 (0100 in binary)
    - " & "
        - Bitwise AND	: Returns 1 if both the bits are 1 else 0.
        ```
        x =           1010
        y =           0100
        x & y =       0000
        
        ```
    - " | "
        - Bitwise OR : Returns 1 if either of the bit is 1 else 0.
        ```
        x =           1010
        y =           0100
        x | y =       1110  = 14(decimal)
        ```
    - " ~ "
        - Bitwise NOT	: Returns oneâ€™s complement of the number.
        ```
        x  =      1010     
        ~x =    -(1010 + 1) 
           =      -(1011) 
           =     -11 (decimal)
        ```
    - " ^ "
        - Bitwise XOR	: Return 0 if both bit are zero or one else 1.
        ```
        Example 1 : 
                x =           1010
                y =           0100
                x ^ y = 14   (1110)
        Example 2 
                x =           1111
                y =           1111
                x ^ y = 0   (0000)
        ```
    - " >> or Bitwise Right Shift"
        - Shifts the bits of the number to the right and fills 0 on left.	Similar effect as of dividing the number with (2 ^ number of shift).
        ```
        a = 10 = 0000 1010 (Binary)
        a >> 1 = 0000 0101 = 5

        Same as a / (2 ^ number of shift)
              
              = 10/2^1
              
              = 10/2 = 5
        ```
    - " << or Bitwise Left Shift"
        - Shifts the bits of the number to the left and fills 0 on right. Similar effect as of multiplying the number with (2 ^ number of shift).
        ```
        a = 5 = 0000 0101 (Binary)
        a << 1 = 0000 1010 = 10

        Same as a * (2 ^ number of shift)
              
              = 5 * 2^1
              
              = 5 * 2 = 10
        ```

````python
a = 10
b = 4
 
# Print bitwise AND operation
print("a & b =", a & b)
 
# Print bitwise OR operation
print("a | b =", a | b)
 
# Print bitwise NOT operation
print("~a =", ~a)
 
# print bitwise XOR operation
print("a ^ b =", a ^ b)
````

````python
a = 10
 
# print bitwise right shift operator
print("a >> 1 =", a >> 1)

b = 5
 
# print bitwise left shift operator
print("a << 1 =", b << 1)
````

**<h2>Membership operators in python</h2>**

- In and not in are the membership operators in Python. They are used to test whether a value or variable is found in a sequence (string, list, tuple, set and dictionary).
    - " in " 	
        - True if value/variable is found in the sequence	
        `5 in x`
    - " not in "	
        - True if value/variable is not found in the sequence	
        `5 not in x`
