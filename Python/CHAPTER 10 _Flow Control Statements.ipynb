**<center><h1>For Loop</h1></center>**

The for loop statement in Python is used to iterate over a sequence data type (list, tuple, string) or other iterable objects.

```
Syntax : 
          for value in sequence:
              statements
```

**Range**

- Using Range() function we can generate a sequence of numbers.

```
Syntax : 
           range(start, end , step)

           Start : Starting value of number (included)
           end : Ending value of number  (excluded)
           step : step defaults to 1
``` 

- This will not store values in memory, it generates values on go. If we want store then we can use list() function.
```
list_sequence_numbers = list(range(0,10))
```
- We will use range function in for loop to iterate over sequence of numbers by providing start and end value. And if want to iterate over sequence data types uisng index then we will provide len() of that sequence data type in the place of end.

```
for value in range(0,10):
    statements

for list_index in range(0, len(list_sequenc)):
    list_value = list_sequenc[list_index]
```

````python
print("Here it will not store into memory and remembers start and end value",
      range(10))

print("if we want to store into a variable we can use list function",
      list(range(10)))
````

````python
# Program to find the sum of all numbers stored in a list
# List of numbers
numbers = [6, 5, 3, 8]

# variable to store the sum
sum = 0

# iterate over the list
for val in numbers:
    sum = sum + val

print("The sum is", sum)
````

````python
# Program to iterate through a list using indexing
genre = ['pop', 'rock', 'jazz']

# iterate over the list using index
for i in range(0, len(genre)):
    print("I like", genre[i])
````

**<center><h1>While Loop</h1></center>**

While loop is used to itereate over block of code until test condition is True.We generally use this loop when we don't know the number of times to iterate beforehand.

```
  Syntax :      
             while test_expression:
                 Body of while
```
  

````python
# Program to add natural
# numbers up to
# sum = 1+2+3+...+n

# To take input from the user,
# n = int(input("Enter n: "))
n = 10

# initialize sum and counter
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i+1 # update counter

# print the sum
print("The sum is", sum)
````

**<center><h1>Python Break and Continue Statements</h1></center>**

In Python, break and continue statements is used to change normal flow of loops. Sometimes we want to terminate loop or skip current iteration without checking test expression.

**Break**

- The break statement is used terminates the loop without checking the condition. Control of the program flows to the statement immediately after the body of the loop.
```
Syntax :     
          for value in sequenc:
              print("terminating loop)
              break
```

**Continue**

- Continue statement is used to skip the remaining lines of code for the current iteration of the loop only. The loop does not end; rather, it continues with the following iteration. 

```
Syntax :     
          for value in sequenc:
              print("terminating loop)
              continue
```

````python
# Use of break statement inside the loop
for val in "string":
    if val == "i":
        break
        
    print(val)
print("The end")
````

````python
# Program to show the use of continue statement inside loops
for val in "string":
    if val == "i":
      print("now skipping")
      continue
    print(val)
    
print("The end")
````

**Pass Statements**

Pass statement in python is a null statement. Interpreter ignores a comment entirely, pass statements is not ignored by interpreter.

If we want to implement function or loop in future, then body of loop or function can not be empty, interpreter will give an error. So, we use the pass statement to construct a body that does nothing.

````python
'''pass is just a placeholder for
functionality to be added later.'''
sequence = {'p', 'a', 's', 's'}
for val in sequence:
    pass


def function(args):
    pass
````
