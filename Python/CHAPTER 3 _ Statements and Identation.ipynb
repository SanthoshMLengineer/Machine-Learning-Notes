**<h1><center>Python Statement, Indentation and comments</center></h1>**

**<h3>Python Statements</h3>**

The instructions that machine can execute is called as statements. Like a = 1 as assignment statements, if statements, for statements etc.

- **MultiLine Statements**

  - Using continuation character `\` we can extend statements into a multiple lines.
  ```
    a = 1 + 2 + 3 + \
        4 + 5 + 6 + \
        7 + 8 + 9
    ```
  - This is an explicit line continuation.
  - In Python, implicit line continuation is implied inside parentheses ( ), brackets [ ], and braces { }. For instance, we can implement the above multi-line statement as
  ```
  a = (1 + 2 + 3 +
       4 + 5 + 6 +
       7 + 8 + 9)

  - Here, the surrounding parentheses ( ) do the line continuation implicitly. Same is the case with [ ] and { }. 
  - For example:
  ```
  colors = ['red',
            'blue',
            'green']

  - We can also put multiple statements in a single line using semicolons, as follows:
  ```
  a = 1; b = 2; c = 3
  ```

**<h3>Python Indentation</h3>**

- Most of programming language like c, c++, java uses braces to define the block of the code.
- In Python uses indentation to define block. A code block starts with indentation and end with unindented block. The amount of indentation must be consistent throughout the block. Generally, four whitespaces are used for indentation and are preferred over tabs. Here is an example.
```
for i in range(1,11):
    print(i)
    if i == 5:
        break 
```

**<h3>Python Comments</h3>**

In python ```#``` symbol is used to start writing the comments. comments helps to describe whats going inside the program.

  ```
  print("Hello") # it prints hello word
  ```

- **MultiLine Comments**

   - We can have comments that extend up to multiple lines. We can use triple quotes, either ''' or """. 
  ```
  """This is also a
perfect example of
multi-line comments"""
```


**<h3>Docstring</h3>**

- Python docstring are string which is used after definition of function, class, modules. Docstring are used to give documentation of the code. We use triple quotation marks for docstring.
```
def square(n):
    '''Takes in a number n, returns the square of n'''
    return n**2
```
- Above ```Takes in a number n, returns the square of n``` is a docstring.

- **Docstring for python function**

  - The docstring for a function or method helps to give short summarize like what it does , and its arguments and return values.
  - We can use `__doc__` attribute to retrieve docstrig.
