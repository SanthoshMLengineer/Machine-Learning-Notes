**<h3>Numbers</h3>**

Integer, floating point numbers and complex numbers are number data type.They are defined as int, float and complex classes in Python. we use type() function to find the type of datatype and we use isinstance() is used to check whether value is having the desired datatype.

```python
# type(value) will give the type of data

print("Printing the type of data type\n")

a = 167
print("{} is a {}".format(a,type(a)))

b = 167.0
print("{} is a {}".format(b,type(b)))

c = 167+2j
print("{} is a {}\n".format(c,type(c)))


# isinstance(value,class) will says True if the value is belongs to
# specified class else
print("checking the that the value is belongs to specified class\n")

a = 167
print("{} is a {} for instance of int".format(a, isinstance(a,int)))

b = 167.0
print("{} is a {} for instance of float".format(b, isinstance(b,float)))

c = 167+2j
print("{} is a {} for instance of complex".format(c, isinstance(c,complex)))
```

**<h3>String</h3>**

String is a sequence of characters. We use single quotes, double quotes to represent and triple quotes to represent multi line strings. String are immutable.

using `+` operator we can **concatenate** two or more strings. We can do **slicing** for string using [ ].

**Methods in string**

- Capitalize
  - This function is used capitalize the string. It gives a replica of the string with its first character capitalized and the rest lowercase.
- find
  - This function is used to find the substring in string. It gives lowest index within the string where substring is found. Return -1 if substring is not found.
- split
  - It is used to split the string using delimiter string as a seperator. It gives a list of the words in the string, using delimiter string. If maxsplit is given, at most maxsplit splits are done. If maxsplit is not specified or -1, then there is no limit on the number of splits all possible splits are made.
- title
  - This function returns titlecased version of the string where words start with an uppercase character and the remaining characters are lowercase.
- upper
  - It gives a replica of string with all uppercase character.
- isalnum
  - The isalnum() method gives result True if all the characters are alphanumeric, meaning alphabet letter (a-z) and numbers (0-9).
  Example of characters that are not alphanumeric: (space)!#%&? etc.
- isalpha
  - The isalpha() method gives result True if all the characters are alphabet letters (a-z).

```python
s = "This is a string"
print(s)

s = '''A multiline
string'''

print(s)
print(s.isalnum())
print(s.isalpha())
print()
```

```python
# slicing
abc = 'Hello world! '


# s[4] = 'o'
print("abc[4] = ", abc[4])

# s[6:11] = 'world'
print("abc[6:11] = ", abc[6:11])

bca = " bangalore"
print("result from concat \n {}".format(abc+bca))

# Generates error Strings are immutable in Python
abc[5] ='d'
```

```python
string = "MACHINE LEARNING"

# Example for capitalize method
print("capitalizing the given string is {}".format(string.capitalize()))

# Example for find method
print("\nprinting starting index of python in given string is {}".\
      format(string.find("python")))

# Example for isalnum method
print("\nchecking is given string is alpha numeric or not result is {}".\
      format(string.isalnum()))

# Example for title method
print("\ntitle case of the string is {}".format(string.title()))

# Example for upper method
print("\nupper case of the string is {}".format(string.upper()))
```
