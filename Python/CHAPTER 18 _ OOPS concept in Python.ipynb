**<h2>Creating a class and object</h2>**

````python
class cars:

    # class attribute
    College_type = "Engineering_college"

    # instance attribute
    def __init__(self, college_name, degree_name):
        self.college_name = college_name
        self.degree_name = degree_name

college_obj = cars("NCET","B.E")

print("Studied {} in {} college".format(college_obj.degree_name,
                                              college_obj.college_name))
````

Lets understand few terminologies
1. self
2. `__init__`
3. Method
4. Class Attribute
5. instance Attribute

- **`self`** : `self` is a special word which is placed as a first parameter in the methods. While calling methods, we will not provide any values to this method. If we call method like myobject.method(arg1, arg2), this is automatically converted by Python into MyClass.method(myobject, arg1, arg2).

- **```__init__```** :  ```__init__``` method is a initialization method, and this method is also called constructor.This method is called automatically when object is created. So when we create object ```car_obj``` then ```__init__``` method called automatically.

- **`Method`** :  `Method` is a type of function defined inside the class defination.

- **`Class variable or class Attribute`**  : A variable that is shared by all instance of class is called class attribute. Class variable defined inside the class but outside of the method.

- **`Instance variable or instance attribute`** : A variable that is defined inside the method. And belongs to only current instance of variable.

We can access the class attribute using ```object_name.attriute_name```. Class attributes are the same for all instances of a class.

````python
print("printing class attribute {}".format(college_obj.College_type))
````

**<h3>Class and objects with methods</h3>**

Methods are nothing but functions defined inside the body of class. Below add_amount, and display_balance are methods.

````python
class bank_account:
    bank_name = "HDFC"

    # instance attributes
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    # instance methods
    def add_amount(self, deposite_amount):
        self.balance = self.balance + deposite_amount

    def display_balance(self):
        print("Balance amount of {} after depositing is {}".format(self.name,
                                                                   self.balance))

account_1 = bank_account("Santhosh", 1000)
account_1.add_amount(1500000)
account_1.display_balance()
````

**<h4>Scope of Variable</h4>**

Variable in OOPS also referred as attributes of class. Depending on life time of varible their are two types
1. Local variable
2. Global variable.

**Local Variables** :
Variable that is defined inside the code block is called Local Variable and that variable can not accessible outside of the block.

Example if a variable defined inside method then it can accessible only inside the block not outside of the code.

````python
# Creates class Car
class Car:
    def start(self):
        message = "Engine started"
        return message


car_a = Car()
print(car_a.message)
````

````python
# Creates class Car
class Car:
    def start(self):
        message = "Engine started"


car_a = Car()
print(car_a.start())
print(car_a.message)
````

**Global Variable**

A global variable is defined outside of any code block and can be accessed anywhere in the class.

````python
# Creates class Car
class Car:
    message1 = "Engine started"

    def start(self):
        message2 = "Car started"
        return message2, self.message1, Car.message1

car_a = Car()
print(car_a.message1)
print(car_a.start())
````

**<h3>Access Modifiers**

Access modifiers is used to modify the scope of variable.

There are three types of access modifiers in Python:
1. public
2. private
3. protected.

- Public access modifiers can be accessed anywhere inside or outside the class. For public variables, you do not have to add any prefixes at all.

- The members of a class that are declared protected are only accessible to a class derived from it. To create a protected variable, you need to prefix a single underscore with the variable name.

- The private variables can only be accessed inside the class. A variable with prefix double underscore is known as private variable.




````python
class Car:
    def __init__(self):
        print ("Engine started")
        self.name = "corolla"
        self.__make = "toyota"
        self._model = 1999
````

````python
car_a = Car()
print(car_a.name)
print(car_a._model)
print(car_a.__make)
````

````python
class student:

    # instance  attribute
    def __init__(self, branch):
        self.branch = branch

    # protected method
    def _display(self):
        print("branch is {}".format(self.branch))


student_obj = student("B.E")
student_obj._display()
````

````python
class student:

    # instance  attribute
    def __init__(self, branch):
        self.branch = branch

    # protected method
    def _display(self):
        print("branch is {}".format(self.branch))


class santhosh(student):

    # instance attribute
    def __init__(self, branch, clg_name):
        student.__init__(self, branch)
        self.clg_name = clg_name

    # public method
    def display(self):
        self._display()
        print("college name {}".format(self.clg_name))

student_obj = santhosh("B.E","NCET")
student_obj.display()

````

````python
class student:

    # instance  attribute
    def __init__(self, branch):
        self.branch = branch

    # protected method
    def _display(self):
        print("branch is {}".format(self.branch))


class santhosh(student):

    # instance attribute
    def __init__(self, branch, clg_name):
        student.__init__(self, branch)
        self.clg_name = clg_name

    # public method
    def display(self):
        student._display(self)
        print("college name {}".format(self.clg_name))

student_obj = santhosh("B.E","NCET")
student_obj.display()

````
