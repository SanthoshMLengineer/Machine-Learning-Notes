**<h1><center>Identifiers and Variables</center></h1>**

**<h2>Keywords</h2>**

Keywords are reserved words in python. These words have specific purpose, so we can not use these words for declaring variable, function etc.


- Those keywords are :
```
from, import, if, else, elif, try, except, raise, finally, break,
pass, continue, await, True, Flase, return, and, or, in, not, for, while,
class, None, is, lambda, as, def, nonlocal, assert, del, global, with,
async, yield
```

**<h2>Identifiers</h2>**

- An identifier is name given to entities like variable, function, class etc. As it helps to differentiate one from the other.


- **Rules for writing identifiers**

  - Identifiers can be a mixture of letters(lowercase or uppercase) or digits or an underscore_.
  - An Identifier can not begin with a digit. 1santhosh is invalid, however santhosh1 is valid identifier. Keywords can not be used as identifiers.
  ```
  lambda = 1 # keywords can't be used throws error
  ```
  - We cannot use special symbols like !, @, #, $, % etc. in our identifier.
  ```
  @santhosh = 1 # we can't use special characters
  ```
  - Python is a case-sensitive language. This means, Variable and variable are not the same.

**<h2>Variables</h2>**

Variables is a named location which is used to store the data.

- For Example :
  ```
  name = "Santhosh"
  ```
- Here we created a variable as name and we stored ```Santhosh``` to it. And we can change it at any time.
  ```
  name = "Santhosh"

  name = "Jagannath"
  ```
- Here initially we stored ```Santhosh``` and we replaced with ```Jagannath``` to it.

- **<h3>Assaigning Values to Variables</h3>**

  - Example 1 : Declare and assign the value. Here we will assign value ```Santhosh``` to ```name``` variable
  ```
  name = "Santhosh"
  ```

  - Example 2 : Here we can assign multiple values to multiple variable in a single statement.
  ```
  a,b,c = 10,20,30
  ```
  - Example 3 : We can assign same value to multiple variables. Below we will be assigning value ```10``` to all three variable (a,b,c).
  ```
  a=b=c=10
  ```

- **<h3>Rules and Naming Convention for Variables</h3>**

  - Constant and variable names should have a combination of letters(lowercase or uppercase) or digits (0 to 9) or an underscore (_).
  - For example:
  ```
  snake_case
  MACRO_CASE
  camelCase
  CapWords
  ```
  - Create a name that makes sense. For example, first_name and second_name  makes more sense than name1 and name2. If variable name having two words, use underscore to separate them.
  - For example:
  ```
  my_name
  current_salary
  ```

**<h2>Python Constants</h2>**

- Constants is a type of variable whose value can not be changed. Usually Constants are declared using capital letters.
- Example 1 : Declare and assign a value.
```
PI = 3.14
```

- **<h3>Rules and Naming Convention for constants</h3>**
  - For example:
  ```
  PI
  G
  MASS
  SPEED_OF_LIGHT
  TEMP
  ```
  - Never use special symbols like !, @, #, $, %, etc.
  - Don't start a variable name with a digit.


**Note**
- Identifier is name given to entity like variable, function, class etc. And variable is name given to memory location.
