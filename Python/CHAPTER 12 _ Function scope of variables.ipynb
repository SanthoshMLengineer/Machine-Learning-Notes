**Global Variables**

Variables declared outside of the function considered to be global variable and it can be accessed inside and outside of the function.

Let's see an example of how a global variable is created in Python

````python
x = "global_variable"

def foo():
    print("x inside:", x)

foo()
print("x outside:", x)
````

In the above code, we created x as a global variable with value ```global_variable``` and printed x value. And defined a foo() to print the global variable x.
Finally, we call the foo() and will print the value of x.

````python
# What if you want to change the value of x inside a function
x = 2

def foo():
    x = x * 2
    print(x)

foo()
````

The output shows an error because Python treats x as a local variable and x is also not defined inside foo(). To make this work, we use the `global keyword`.

**What is the global keyword**

In Python, global keyword allows you to modify the global variable inside the function. 

````python
x = 2

def foo():
    global x
    x = x * 2
    print(x)

foo()
````

From above example we are accessing the global variable inside the fucntion using `global keyword`

**Global variable in Nested Functions**

We can declare global variable inside the function using `gloabl keyword` and we can access that global variable outside the function.

````python
def local_function():
    x = 20

    def global_function():
        # Declaring global variable inside the function
        global x
        x = 25
    print("Before calling bar: ", x)
    global_function()
    print("After calling bar: ", x)

local_function()
print("Global variable x in main: ", x)
````

- In the above program we declared local variable `x = 20` in local_function.
- In global_function we declared global variable using global keyword
``` 
global x 
x = 25
``` 
- we can access and modify `x` outside of the function as it is declared as global variable in global_function.

- Inside local_function, x has no effect of the gloabl keyword. Before and after calling global_function(), the variable x takes the value of local variable i.e x = 20. Outside of the local_function() function, the variable x will take value defined in the global_function() function i.e x = 25. 
- This is because we have used global keyword to create x global variable
inside the global_function() function (local scope). If we make any changes inside the global_function() function, the changes appear outside the local scope, i.e. local_function().

**Local Variables**

A variable declared inside the function's body is known as a local variable

````python
def foo():
    y = "local"
    print(y)
    
foo()
````

**Scope and Lifetime of variables**

The lifetime of variables within a function is as long as the function executes. They are destroyed once we return from the function. Hence, a function does not remember the value of a variable from its previous calls. 

````python
def my_func():
    x = 10
    print("Value inside function:",x)

    
x = 20
my_func()
print("Value outside function:",x)
````

Here we declared gloabl variable outside of the function `x = 20`. In my_func() declared `x = 10` as local variable. In my_func() the value will be printed as 10 as it is local variable.

This `x = 10` value will be destroyed once we return from the function. This is because the variable x inside the function is local scope. On the other hand, variables outside of the function are visible from inside. They have a global scope.

We can read these values from inside the function but cannot change them.
