**<h3><center>IF ELSE STATEMENTS</center></h3>**

If we want execute block of code only if certain condition met then we can use if..elif..else statements..

**If Statement**

If ```if condition``` statements is True then all block under if statements will be executes. If ```if condition``` statements is False then it skips the blocks under the if statements.

````python
# If the number is positive, we print an appropriate message
num = 3
if num > 0:
    print(num, "is a positive number.")


num = -1
if num > 0:
    print(num, "is a positive number.")
````

**If-else Statements**

In ```if-else``` Statements , checks the condition of ```if statements``` if it is true then block under the if statements will be executed. If ``` if condition``` evaluated as False then else block will executed.

````python
if True:
    print("hi")
else:
    print("False")
````

````python
num = 3

if num >= 0:
    print("Positive or Zero")
else:
    print("Negative number")
````

**Elif**

- If we have more than two blocks of code and different condition to be check then ```elif``` will be used.

````python
'''In this program,
we check if the number is positive or
negative or zero'''
num = 3.4
# Try these two variations as well:
# num = 0
# num = -4.5
if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
````

**Python Nested if statements**

We can have a if...elif...else statement inside another if..elif...else statements. Indentation is the only way to figure out the level of nesting.

````python
'''In this program, we input a number
check if the number is positive or
negative or zero'''

num = float(input("Enter a number: "))

if num >= 0:
  if num == 0:
      print("Zero")
  else:
      print("Positive number")
else:
    print("Negative number")
````
