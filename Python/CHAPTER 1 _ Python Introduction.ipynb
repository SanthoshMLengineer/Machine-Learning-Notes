**<center><h1>Python</h1></center>**

Python language created by Van Guido Rossam. It is general purpose and widely used programming language. It is easy to learn and easy to understand.

**<h2>Characterstics of Python</h2>**

- It supports both object and functional oriented programming language.

- It is dynamically typed programming language, you no need to specify the type of data giving to the varible.

- It supports automatic garbage collection.

- It is interpreted programming language.

**<h2>Application</h2>**

- It is used in web development, Android Application Development, Artificial Intelligence technologies, data analysis, web scrapping.
