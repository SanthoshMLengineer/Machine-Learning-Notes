**<center><h1>Function Arguments</h1></center>**

Function can have any number of parameters and it is optional. In program below function greeting() has two parameters `name and msg`. It will run smoothly if we call this function with two arguments otherwise it will give error.

````python
def greeting(name, msg):
    """This function greets to
    the person with the provided message"""
    print("Hello", name + ', ' + msg)


greeting("Santhosh", "Good morning!")
````

````python
greeting() # here we get error because we are not passing arguments to the greet function
````

````python
"""
here we get error because we are passing only one paramter
"""
greeting("Santhosh")
````

**Function Default Parameter**

- Function can have default arguments which we can assign default values using assignment operator (=).

````python
def greeting(name, msg="Good morning!"):
    """
    This function greets to
    the person with the
    provided message.
    If the message is not provided,
    it defaults to "Good
    morning!"
    """
    print("Hello", name + ', ' + msg)

greeting("SANTHOSH")

greeting("Jagannath", "How do you do?")
````

-  In this function `msg` has a default value, so it is optional whereas `name` parameter does not have a default value , so it is required parameter during a call.
-  Non-default arguments cannot follow default arguments.

**Python Keyword Arguments**

- When we call a function with some values with coresponding arguments then these type of arguments called keyword arguments.
```
# Keyword arguments
greeting(name = "Bruce",msg = "How do you do?")
# without keyword arguments
greeting("Bruce", "How do you do?")
```

-  When we called it as greeting("Santhosh", "How are you?") , the value "Santhosh" gets assigned to the argument `name` and similarly "How are you?" to `msg`. And when we call using keyword arguments then the order (position) of the arguments is not important and it can be changed.

````python
# 2 keyword arguments
greeting(name = "Bruce",msg = "How do you do?")

# 2 keyword arguments (out of order)
greeting(msg = "How do you do?",name = "Bruce")

# 1 positional, 1 keyword argument
greeting("Bruce", msg = "How do you do?")
````

- We must keep in mind that keyword arguments must follow positional arguments otherwise interpreter will give errors. For example, the function call as follows:

````python
greeting(name="Bruce", "How do you do?")
````

**Python Arbitrary Arguments**

The number of arguments that will be supplied into a function may not always be known in advance. Python allows us to handle this kind of situation through function calls with an arbitrary number of arguments. In the function definition, we use an asterisk (*) before the parameter name to denote this
kind of argument.

````python
"""
Here, we have called the function with multiple arguments.
These arguments get wrapped up into a tuple before being passed into the
function. Inside the function, we use a for loop to retrieve all the
arguments back.
"""

def greet(*names):
    """This function greets all
    the person in the names tuple."""
    # names is a tuple with arguments
    for name in names:
        print("Hello", name)


greet("Santhosh", "Vedant", "Shekhar", "Tony","Srikant")
````

`*` is for storing positional arguments in tuple and `**` is used to store keyword arguments in dictionary format.
