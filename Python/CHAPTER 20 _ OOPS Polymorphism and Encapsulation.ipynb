**<h1>Polymorphism**

Polymorphism is nothing but having many forms. Methods in class can have same names with different parameters,and child class can have same method name as parent.

Polymorphism can be implemented in two ways:
1. Method Overloading
2. Method Overriding

**Method Overloading**

Method in class can behave different ways depending on the number and type of parameters.

```python
# Creates class Car
class Car:
   def start(self, a, b=None):
        if b is not None:
            print (a + b)
        else:
            print (a)
```

```python
car_a = Car()
car_a.start(10)
```

```python
car_a.start(10,20)
```

**Method Overriding**


Method overriding refers to having a method with the same name in the child class as in the parent class. Definition in child class and parent class differs but having same name. Child class method will override the method of parent class.

```python
class bank:

    # __init__ is known as the constructor
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def details(self):
        print("My name is {}".format(self.name))
        print("balance: {}".format(self.balance))

# child class
class employee(bank):
    def __init__(self, name, balance, branch):
        self.branch = branch

        # invoking the __init__ of the parent class
        bank.__init__(self, name, balance)

    def details(self):
        print("My name is {}".format(self.name))
        print("Balance: {}".format(self.balance))
        print("branch: {}".format(self.branch))


# creation of an object variable or an instance
a = employee('Rahul', 886012, "CSE")

a.details()
```

**<h2>Encapsulation</h2>**

Encapsulation is nothing but hiding the data. We can restrict access to methods and variables.we denote private attributes using underscore as the prefix i.e single _ or double __.



```python
class Computer:

    def __init__(self):
        self.__maxprice = 900

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price


# __maxprice is private data so we can not access it.
c = Computer()
c.__maxprice
```

```python
class Computer:

    def __init__(self):
        self.__maxprice = 900

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price


# __maxprice is private data so we can not access it.
c = Computer()
c.setMaxPrice(1000)
c.__maxprice
```
