**<center><h1>Dictionary Data Types</h1></center>**

Python dictionary is a mapping data type which stores values in key/value format. Values can be retrived using keys.

Dictionary can be created by placing `key:value` pair inside curly braces {} separated by commas. Key and value are seperated by colon(:). Values can be any data type and can repeat but keys must be unique and immutable type(string, numbers, tuple).
  ```
  # Initialization with no elements
  dict_one = dict()

  # Initialization with elements
  dict_two = {1:"one", 2:"two"}
  ```

**Accessing Elements from Dictionary**
  
- Dictionary uses keys to access values. Keys can be used either inside square brackets `[ ]` or get() method to access values
- Using `[ ]` to access values using keys we will get keyerror if key is not found. get() method returns None if the key is not found.

**Changing and Adding Dictionary elements**

- We can add `key:value` pair to dictionary, if key is present in dictionary then value will be updated and if key is not present new `key:value` pair will be added.

**Removing elements from Dictionary**

- The pop() function is used to delete a particular item from a dictionary.
With the supplied key, this method removes the item and returns the value.
- All items can be removed at once using the clear() function. The del keyword can also be used to delete the all items or individual items.

**Methods in Dictionary**

- keys()
  - Returns dictionary's keys.
- values()
  - Returns dictionary's values.
- items()
  - Return dictionary's items in (key, value) format.
- update()
  - The update() method adds element(s) to the dictionary if the key is not in the dictionary. If the key is in the dictionary, it updates the key with the new value.

````python
# empty dictionary
my_dict = {}

# dictionary with integer keys
my_dict = {1: 'Machine Learning', 2: 'NLP'}
print("\nDictionary with integer keys",my_dict)

# dictionary with mixed keys
my_dict = {'Course': 'Machine Learning', 1: [2, 4, 3]}
print("\nDictionary with mixed keys",my_dict)

# using dict()
my_dict = dict({1: 'Machine Learning', 2: 'NLP'})
print("\nDictionary initialization",my_dict)

# from sequence having each item as a pair
my_dict = dict([(1,'Machine Learning'), (2,'NLP')])
print("\nDictionary initialization",my_dict)
````

````python
# get vs [] for retrieving elements
my_dict = {'Course': 'Machine Learning', 'Salary': "15LperYear"}

# Output: Jack
print(my_dict['Course'])

# Output: 26
print(my_dict.get('Salary'))

# Trying to access keys which doesn't exist throws error
# Output None
print(my_dict.get('address'))

# KeyError
print(my_dict['address'])
````

````python
# Changing and adding Dictionary Elements
my_dict = {'Course': 'Machine Learning', 'Salary': "15LperYear"}

# update value
my_dict['Salary'] = "25LperYear"
#Output: {'age': 27, 'name': 'Jack'}
print(my_dict)

# add item
my_dict['Tutorials'] = 'SanthoshAITutorials'
# Output: {'address': 'Downtown', 'age': 27, 'name': 'Jack'}
print(my_dict)
````
