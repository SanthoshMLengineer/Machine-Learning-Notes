**<h1><center>Inheritance</center></h1>**

Inheritance is nothing but one class can inherit the properties of another class. New class is called derived class or child class and old class or existing class is called base class or parent class.


Types of Inheritance

1. Single Inheritance :  In this type of inheritance derived class can only inherit the properties from only one parent class.

<center><img src = "https://media.geeksforgeeks.org/wp-content/uploads/Single.jpg" width = 150px></center>

2. Multilevel Inheritance : In this type of inheritance one class inherits the properties of derived class and which in tern inherits the properties of parent class.

<center><img src = "https://www.cdn.geeksforgeeks.org/wp-content/uploads/multilevel-inheritance.jpg" width = 150px></center>

3. Hierarchical Inheritance : In this type of inheritance one base class can have more than one derived class.

<center><img src = "https://www.programtopia.net/wp-content/uploads/2021/01/heirar_1.png" width = 300px></center>

4. Multiple Inheritance : A derived class can inherits more than one base class properties.

<center><img src = "https://media.geeksforgeeks.org/wp-content/uploads/Multiple-1-1.jpg" width = 200px></center>

````python
class Parent:
    def parent_method(self):
        print("Parent method")

class Child(Parent):
    def child_method(self):
        print("Child method")

child_obj = Child()
child_obj.parent_method()  # Accessing parent method
````

````python
class bank:

    # __init__ is known as the constructor
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance


# child class
class employee(bank):
    def __init__(self, name, balance, branch):
        self.branch = branch

        # invoking the __init__ of the parent class
        bank.__init__(self, name, balance)

    def details(self):
        print("My name is {}".format(self.name))
        print("balance: {}".format(self.balance))
        print("branch: {}".format(self.branch))


# creation of an object variable or an instance
a = employee('Santhosh', 886012, "Bangalore")
a.details()
````
