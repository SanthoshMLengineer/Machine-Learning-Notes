<b><u>Sub Query</u></b>


A query is nested with another sql query is known as sub query.

Their are three types of sub query that Scaler sub query, Multiple row sub-query, Correlated sub-query.

<u>Scaler sub-query</u>:
It returns only one row and one column, it means returns only single value.
```
Write a query to return all employees who salary is greater than the avarage salary of all employees.

SELECT *
FROM employees as e
where salary > (select avg(salary) from employees);
```

<u>Multiple row sub-query</u>:

It means that it returns either multiple row and multiple column or multiple row and single column.

```
Find the employees who's salary is highest salary in each department

select *
from eployee
where (dept_name, salary) IN(select dept_name, MAX(salary)
                             from employee as e
                             groupby dept_name)
```


<u>Correlated sub-query</u>

It means that query is correlated to outer query.A correlated subquery depends on the outer query and executes once for each row processed by the outer query. It can access columns from the outer query.

It runs sub-query as any times outer query runs.

```
SELECT
    column1
FROM
    table1 t1
WHERE
    column2 > (SELECT AVG(column2) FROM table2 WHERE t1.id = table2.id);
```
