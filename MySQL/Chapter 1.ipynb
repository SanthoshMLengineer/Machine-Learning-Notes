<u>Database</u>

Databases use a systematic approach for storing data, typically organized into tables, rows, and columns within a database management system.

<u>Data base management system</u>

DBMS is a software which is used to create, organize and manage the databases.

<u>Relational DBMS</u>

Relational DBMS is a type of DBMS where it works on the concept of relational model which means that data is organized into tables with rows and columns. Tables have predefined relationships, and data integrity is maintained through constraints.

RDBMS systems typically use SQL (Structured Query Language) as the standard query language for interacting with the database.

<u>SQL and MySQL</u>

Structured query language is language is used to interact with the relational databases.And MySQL is RDBMS which used SQL to interact with it.



We used SQL to do CRUD operations:
1. Create : Create databases, tables
2. Read : Used to read the data
3. Update : Used to update the existing data.
4. Delete : Used to delete database, tables or specific row.


<u>SQL data types</u>

Data types tells what type of data we are storing in a table or variable.

1. String Data Type :
  
  - CHAR : Fixed length character string with specified length 'n'. When you define a column as CHAR(n), it will always reserve a fixed length of 'n' characters, padding with spaces if the actual data is shorter than 'n'.
  - VARCHAR : Variable length charcter string with maximum specified length 'n'. When you define a column as VARCHAR(n), it will only use the necessary amount of storage to store the actual data, without padding.
  - TEXT : Variable length character string with no maximum length specified.

2. Integer Data Type :

  - INT or INTEGER : It typically store integer with 32 bit signed integer
  - SMALLINT : It typically represents a 16-bit signed integer.
  - BIGINT : It typically store integer with 64 bit signed integer.
  - FLOAT : It is used to store the floating point values with specified precision values.It is typically represented in 64 bit.
  - REAL : It is used to store single-precision floating-point numbers.(32 bit)

3. Boolean Data type :    
  - BOOLEAN or BOOL: Represents true or false values.

4. DATE and Time Data Type :    
  - DATE: Date values in the format 'YYYY-MM-DD'.
  - TIME: Time values in the format 'HH:MI:SS'.
  - DATETIME or TIMESTAMP: Combined date and time values.

<u>Types of SQL commands</u>

1. Data Query Language (DQL) : In this commands are used to read or retrieve data from one or more tables.
2. Data Definition Language(DDL) : In this commands used for
  - CREATE: Creates a new database object (table, view, index, etc.).
  - ALTER : Modifies the structure of an existing database object.
  - DROP: Deletes an existing database object.

3. Data Manipulation Language(DML) : Used to modify the database.
  - INSERT: Adds new records into a table.
  - UPDATE: Modifies existing records in a table.
  - DELETE: Removes records from a table based on a condition.
4. Data Control Language(DCL) : Used to grant & revoke permissions.
  - GRANT: Provides specific privileges to users or roles.
  - REVOKE: Removes specific privileges from users or roles.

5. Trasaction control language(TCL) : Used to manage transactions.
  - COMMIT: Saves all changes made during the current transaction.
  - ROLLBACK: Undoes all changes made during the current transaction.
  -SAVEPOINT: Sets a point within a transaction to which you can later roll back.
  -ROLLBACK TO SAVEPOINT: Rolls back to a specified savepoint.
