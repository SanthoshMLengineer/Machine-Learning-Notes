**<center><h1>Descriptive Statistics</h1></center>**

In any data science projects, initial step is data cleaning and data analysis. And this is most important step. To start data analysis part, some description about data is neccessary like number columns, types of data and some stastical information.

- `head` function is used to display data. It will display only 5 rows as default, if we want more rows to display then we can put in head function like `head(10)` 10 rows are displayed.

- `info` function is used to give some basic information about data like number of columns, type of data and memory used.

```python
import pandas as pd
import numpy as np

#Create a Dictionary of series
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack',
                       'Lee','David','Gasper','Betina','Andres']),
     'Age':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
     'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,
                         3.78,2.98,4.80,4.10,3.65])
}

#Create a DataFrame
df = pd.DataFrame(d)

print(df.head())
```

```python
df.select_dtypes(exclude='object')
```

```python
df.select_dtypes(include='object')
```

```python
df.head(10)
```

```python
df.info()
```

info function provides basic information.Their are 3 columns Name. Age, Rating where Name and Rating are categorical column and Age is numerical column.And memory usage is memory usage: 416.0+ bytes.

```python
df.describe()
```

This function gives the mean, std and IQR values. And, function excludes the character columns and given summary about numeric columns. 'include' is the argument which is used to pass necessary information regarding what columns need to be considered for summarizing. Takes the list of values; by default, 'number'.

        object − Summarizes String columns
        number − Summarizes Numeric columns
        all − Summarizes all columns together (Should not pass it as a list value)

```python
print(df.describe(include=['object']))
```

```python
print(df. describe(include='all'))
```
