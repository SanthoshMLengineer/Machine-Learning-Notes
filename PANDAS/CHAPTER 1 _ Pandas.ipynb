
**<center><h1>Pandas</h1></center>**

Pandas is widely used data analysis and data manipulation tool developed by wes MCkinney.

It is built on top of Numpy package and it has popular data structure is called DataFrame. DataFrame allows you to store and modify tabular data.

**Key Features of Pandas**

- Pandas has different types of data structure that is series and DataFrame which helps to store and manipulate data.
- Helps to handle missing data.
- Label-based slicing, indexing.
- Columns from a data structure can be deleted or inserted.


Pandas has two types of data structures :

- Series

- DataFrame

**Series**
- Series is one-dimensional array which can have only homogenous data.
- Key Points
    - Homogeneous data
    - Size Immutable
    - Values of Data Mutable


**DataFrame**
- DataFrame is a two-dimensional array which can have heterogeneous data. For example,

          | Name    | Age    | Gender    | Rating    |
          |---------|--------|-----------|-----------|
          |   Steve |  32    |   Male    | 3.45      |
          |---------|--------|-----------|-----------|
          |   Lia   |  28    |   Female  | 4.6       |
          |---------|--------|-----------|-----------|
          |   Vin   | 45     |   Male    | 3.9       |
          |---------|--------|-----------|-----------|
          |   Katie | 38     |   Female  | 2.78      |


The data is represented in rows and columns.

- Key Points
    - Heterogeneous data
    - Size Mutable
    - Data Mutable
