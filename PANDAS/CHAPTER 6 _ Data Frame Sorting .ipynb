**<h1><center>Sorting</h1></center>**

Pandas has inbuilt function to sort index and values.

There are two kinds of sorting available in Pandas they are :
1. By label
2. By Actual Value

```python
"""
Unsorted Data Frame
In unsorted_df, the labels and the values are unsorted.
Let us see how these can be sorted.
"""
import pandas as pd
import numpy as np

unsorted_df = pd.DataFrame(np.random.randn(10,2),
                           index=[1,4,6,2,3,5,9,8,0,7],
                           columns=['col2','col1'])

print(unsorted_df)
```

**<h3>1. By Label</h3>**

`sort_index()` method is used to sort dataframe. By passing axis arguments and order of sorting, helps to define the way of sorting.

<u>`axis arguments`</u> : `axis arguments` accepts values either `0 or 1` . If we provide value `0` then it sorts by the index and if we provide value `1` then column will be sorted. By default, axis=0, sort by index.

<u>`ascending parameter`</u> : By passing the Boolean value to ascending parameter, the order of the sorting can be controlled. Let us consider the following example to understand the same.

```python
"""
ascending parameter helps to sort the values either ascending or descending
ascending = True means the values sorted in ascending
ascending = False means the values sorted in descending
"""
unsorted_df = pd.DataFrame(np.random.randn(10,2),
                           index=[1,4,6,2,3,5,9,8,0,7],
                           columns = ['col2','col1'])

sorted_df = unsorted_df.sort_index(ascending=False)
print(sorted_df)
```

```python
"""
This one helps to demonstrate the column sorting
axis = 1 means column sorting

axis = 0 means row sorting
"""
unsorted_df = pd.DataFrame(np.random.randn(10,2),
                           index=[1,4,6,2,3,5,9,8,0,7],
                           columns = ['colb','cola'])

print(unsorted_df)
sorted_df = unsorted_df.sort_index(axis=1)

print(sorted_df)
```

```python
"""
This one helps to demonstrate the column sorting
axis = 1 means column sorting

axis = 0 means row sorting
"""
unsorted_df = pd.DataFrame(np.random.randn(10,2),
                           index=[1,4,6,2,3,5,9,8,0,7],
                           columns = ['colb','colB'])

print(unsorted_df)
sorted_df = unsorted_df.sort_index(axis=1)

print(sorted_df)
```

```python
"""
This one helps to demonstrate the column sorting
axis = 1 means column sorting

axis = 0 means row sorting
"""
unsorted_df = pd.DataFrame(np.random.randn(10,2),
                           index=[1,4,6,2,3,5,9,8,0,7],
                           columns = ['colb','colB'])

print(unsorted_df)
sorted_df = unsorted_df.sort_index(axis=1, ascending = False)

print(sorted_df)
```

**<h3>By Value</h3>**

`sort_values()` is the method for sorting by values. It accepts a `by` argument which will use the column name with which the values are to be sorted.

```python
import pandas as pd
unsorted_df = pd.DataFrame({'col1':[2,1,1,1],
                            'col2':[1,3,2,4]})

sorted_df = unsorted_df.sort_values(by='col1', ascending=False)

print("unsorted df \n",unsorted_df)
print("sorted by col1\n",sorted_df)
sorted_df.iloc[1,0]
```

```python
data = {'SecondHighestValue': [sorted_df.iloc[1,0]]}

df = pd.DataFrame(data)
df
```

```python
unsorted_df.sort_values(by=['col1','col2'])
```

First, it will sort based on values in `col1` column. Since the values `1` in column `col1` is repeated then it will sort again by values in `col2` in ascending order.

```python
unsorted_df = pd.DataFrame({'col1':[2,1,0,4],
                            'col2':[1,3,2,4]})

sorted_df = unsorted_df.sort_values(by=['col1','col2'])

print("unsorted df \n",unsorted_df)
print("sorted by col1\n",sorted_df)
```

**Sorting Algorithm**

We can choose sorting algorithm by passing type of algorithm to `kind` argument.

```python

sorted_df = unsorted_df.sort_values(by='col1' ,kind='mergesort')

print(sorted_df)
```
