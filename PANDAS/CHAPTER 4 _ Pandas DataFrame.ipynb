**<h3>Column Selection, addition, deletion</h3>**
<u>**Column Selection**</u> :   We can select the column by placing column name in square brackets along with dataframe name.

`
Syntax : dataframe['column_name']
`


````python
import pandas as pd

d = {'one' : pd.Series([1, 2, 3],
                       index=['a', 'b', 'c']),
    'two' : pd.Series([1, 2, 3, 4],
                      index=['a', 'b', 'c', 'd'])}

df = pd.DataFrame(d)
print(df)
print("\nColumn one is selected\n",df['one'])
````

<u>**Column Addition**</u> :
`Syntax : dataframe[column_name] = data`

Column_name is added to dataframe if column_name is not present in the  column list of dataframe.

Values will be updated if column_name is present in the column list of dataframe

````python
# we will work on above dataframe.
print(df)

# Adding a new column to an existing DataFrame
# object with column label by passing new series

print ("Adding a new column by passing as Series:")
df['three'] = pd.Series([10,20,30],
                        index=['a','b','c'])

print("\n",df)
````

````python
print ("Adding values for existing columns.")

df['two'] = pd.Series([10,20,30,40],
                        index=['a','b','c','d'])

print(df)
````

````python
print ("Adding values for existing columns.")

df['two'] = pd.Series([10,20,30],
                        index=['a','b','c'])

print(df)
````

<u>**Column Deletion**</u> : Columns can be deleted using dataframe functions `del or pop or drop`.

`drop` function as an axis argument it takes input either `0 or 1`. `0` indicates rows and `1` indicates columns.

````python
print(df)

# using del function
print ("Deleting the first column using DEL function:")
del df['one']
print(df)

# using pop function
print ("Deleting another column using POP function:")
df.pop('two')
print(df)

# using drop function
print ("Deleting another column using drop function:")
df.drop(['three'], axis  = 1, inplace =   True)
````

````python
print(df)
````

**<h3>Row Selection, Addition, and Deletion</h3>**

We will now understand row selection, addition and deletion through examples. Let us begin with the concept of selection.

<u>**Selection by Label**</u> : Rows can be selected by passing row label to a loc function.

`Syntax : dataframe.loc[[row label],[column label]]`.

loc function is used access rows and columns by providing row label and column labels of dataframe.

````python
import pandas as pd

d = {'one' : pd.Series([1, 2, 3],
                       index=['a', 'b', 'c']),
   'two' : pd.Series([1, 2, 3, 4],
                     index=['a', 'b', 'c', 'd'])}

df = pd.DataFrame(d)
print(df)
print(df.loc['b':'d',['one']])
````

````python
import pandas as pd

d = {'one' : pd.Series([1, 2, 3],
                       index=[0,1,2]),
   'two' : pd.Series([1, 2, 3, 4],
                     index=[0,1,2,3])}

df = pd.DataFrame(d)
print(df)
print(df.loc[0:2,['one']])
````

````python
df.loc[df['one']>1, ['one','two']]
````

<u>**Selection by integer location**</u> : Rows can be selected by passing index of rows and columns to an iloc function.

````python
df = pd.DataFrame(d)

print(df.iloc[2])
print(df.iloc[0:2,[0]])
````

````python
print(df.iloc[0:2,['one']])
````

````python
print(df.iloc[df['one']>1,[0]])
````

**Difference between loc and iloc?**

<u>Loc function</u> : Loc function is used to access rows and columns by providing rows and column labels. It uses labels to access elements of dataframe.

If row label or column label is not present in the dataframe it gives keyerror.

<u>iloc function</u> : iloc function is also used to access row and columns by providing row and column index. It uses index to access elements of dataframe.

If row and column index is not present in the dataframe then it will provide indexerror.

**<h3>Addition of Rows</h3>**

<u>Append function</u> : Append function is used to add new rows at the end.

```
syntax : dataframe1.append(data,
                          ignore_index = =True)

        data : dictionary or Pandas Series or Dataframe
        ignore_index :  True Means the index from the series or the source dataframe will be ignored.
                        The index available in the target dataframe will be used, instead.
                        sFalse means otherwise.
```

````python
import pandas as pd

df = pd.DataFrame([[1, 2], [3, 4]],
                  columns = ['a','b'])

df2 = pd.DataFrame([[5, 6], [7, 8]],
                   columns = ['a','b'])

df = df.append(df2, ignore_index  =   True)
print(df)
````

````python
import pandas as pd

df = pd.DataFrame([[1, 2], [3, 4]],
                  columns = ['a','b'])

df2 = pd.DataFrame([[5, 6], [7, 8]],
                   columns = ['c','b'])

df = df.append(df2, ignore_index  =   True)
print(df)
````

**<h3>Deletion of Rows</h3>**

Use index of row to drop the rows from a dataframe. If index is duplicated, then multiple rows will be dropped.

````python
import pandas as pd

df = pd.DataFrame([[1, 2], [3, 4]], columns = ['c','d'])
df2 = pd.DataFrame([[5, 6], [7, 8]], columns = ['c','d'])

df = df.append(df2)

print(df)
# Drop rows with label 0
df = df.drop(0)

print("Data frame after dropping\n",df)
````
