**<center><h1>Data Visualization</h1></center>**

After data collection part is done, next step is to understand, analyze the data patterns, trends and relationship between variables in the data. Different Data plots helps us to get insights about the data. Their are several packages used for plotting the data. The most popularly used packages are Matplotlib and seaborn.

Their are few important plots:

1. Bar Graph
2. Histogram
3. Pie Chart
4. Box Plot
5. Line Chart
6. Scatter Plot


**<h3>Bar Graph<h3>**

Bar graph works on categorical data. This graph helps to compare quantity of different categories present in the same column. It displays categories in rectangular box whose height is determined by how many times each category is present in the column.

  <center><img src = "https://www.mathsisfun.com/data/images/bar-graph-fruit.svg"></center>

- The above graph presents that category Apple has occured 35 times in the column similarly 30 times Orange, 10 times Banana, 25 times Kiwifruit, 40 times, Blueberry, 5 times Grapes. From that we can conclude that Blueberry has high occurences and Grapes has low occurence.

**<h3>Pie Chart</h3>**

This is suitable for categorical data where each category represents a portion or percentage of the total.

  ```
  from matplotlib import pyplot as plt
  import numpy as np
  fig = plt.figure()
  ax = fig.add_axes([0,0,1,1])
  ax.axis('equal')
  langs = ['C', 'C++', 'Java', 'Python', 'PHP']
  students = [23,17,35,29,12]
  ax.pie(students, labels = langs,autopct='%1.2f%%')
  plt.show()
  ```

<center><img src="https://www.tutorialspoint.com/matplotlib/images/pie_chart.jpg" width = "350px"></center>

From above image gives ```14.66%``` for label ```c++```. Lets calculate the percentage
  ```
  c = 17/23+17+35+29+12
    = 17/116
    = 0.146551
  ```


**<h3>Box Plot</h3>**

The box plot is also known as whisker plot, this plot helps to describe some information about continuous data like minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.


<center><img src='https://www.tutorialspoint.com/matplotlib/images/box_plot.jpg' width = 500px></center>


**<h3>Line Chart</h3>**

This graph helps to track changes over both short and long peroid of time. This graph helps to get whether two set of values is dependent on each other or not. Line graph has a horizontal axis called the x-axis and a vertical axis called the y-axis. The x-axis usually has a time period over which we would like to measure the quantity of a specific thing or an item in the y-axis. A line graph is used to assess whether the quantity on the y-axis is rising or decreasing over time.
