**<h1><center>DATA VISUALIZATION USING MATPLOTLIB</center></h1>**

Matplotlib is a data visualization tool used to create 2D graphs.It has a module called pyplot which helps to create graphs by providing various properties of graph like style, color, size of graph etc.

we will import pyplot module from matplotlib to create different types of plots```import matplotlib.pyplot as plt```

Important points  to note :

- ```plt.plot()```for plotting line chart similarly in place of plot other functions are used for plotting. All plotting functions require data and it is provided in the function through parameters.
- ```plt.xlabel , plt.ylabel``` for labeling x and y-axis respectively.
- ```plt.xticks , plt.yticks``` for labeling x and y-axis observation tick points respectively.
- ```plt.legend()``` for signifying the observation variables.
- ```plt.title()``` for setting the title of the plot.
- ```plot.show()``` for displaying the plot.

**Histogram**

This plot is used to check the distribution of numerical data.
```
Syntax:
      matplotlib.pyplot.hist(x,
                            bins=None,
                            edgecolor = None,
                            facecolor = None
                            )

      x : This parameter are the sequence of data.
      bins : This parameter tells how many number of bins should be their in graph.
      edgecolor : Border color
      facecolor : Bins color

```

````python
import numpy as np

import matplotlib.pyplot as plt

x = np.random.randn(10000, 3)

plt.hist(x,
         color = ["green","blue","black"],
         bins = 20)

plt.legend(["green","blue","black"])

plt.xlabel("X - axis")

plt.xlabel("Y - axis")

plt.title("Histogram")

plt.show()
````

**Bar Graph**

This plot is used to check distribution of categorical data.

```
Syntax :
    matplotlib.pyplot.bar(x,
                          height,
                          width,
                          color)

    x : lIst or series which should be categorical.
    height : numerical values which tells the height of each category.
    width : width of bar.
    color : color of bar.
```


````python
dict_marks_subject = {"maths" : 89,
                      "science" : 76,
                      "computer" : 82}

x = dict_marks_subject.keys()

height = dict_marks_subject.values()

plt.bar(x,
        height)
````

**<h3>Pie Graph</h3>**

Data points in the series will be shown as percentage of whole pie.

```
Syntax
      matplotlib.pyplot.pie(x,
                            explode=None,
                            labels=None,
                            colors=None,
                            autopct=None)

      x : array
      labels : labels for each value in numerical data
      colors : A sequence of matplotlib color args through which
      the pie chart will cycle. If None, will use the
      colors in the currently active cycle.
      autopct : Default None,
                If not None, is a string or function used to label
                the wedges with their numeric value.

    
```

````python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.pie(students,
        labels = langs,
        autopct='%1.2f%%')
````

**<h3>Box Plot</h3>**

This plot is used to summarise continuous data.It gives mean, median, mode, 25%, 50% etc.

```
Syntax :  matplotlib.pyplot.boxplot(data,
                                    widths=None)

          data : array or sequence of array to be plotted

          width : Sets the width of each box either with a scalar
          or a sequence. The default is 0.5
```

````python
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

# Creating plot
plt.boxplot(data)
````

**<h3>Line Plot</h3>**

This graph helps to track changes over both short and long peroid of time. Line graph has a horizontal axis called the x-axis and a vertical axis called the y-axis.


```
Syntax :
          matplotlib.pyplot.plot(x,
                                 y,
                                 linestyle)
          x, y : array-like or scalar
          
          The horizontal / vertical coordinates of the data points.
          x values are optional and default to range(len(y)).

          linestyle : '-'  solid line style
                      
                      '--'  dashed line style
                      
                      '-.'  dash-dot line style
                      
                      ':'   dotted line style'-'
    

```

````python
# define data values
x = np.array([1, 2, 3, 4])  # X-axis points
y = x*2  # Y-axis points

plt.plot(x, y)  # Plot the chart
plt.show()  # display
````

**<h3>Scatter Plot</h3>**

Scatter plot helps us to analyse the relationship between two variables.Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show relationship between two variables.

```
Syntax :    

        matplotlib.pyplot.scatter(x,
                                  y,
                                  linewidths)

        x , y : Arrays should have same length.
        linewidths- width of marker border
```

````python
x =[5, 7, 8, 7, 2, 17, 2, 9,
    4, 11, 12, 9, 6]

y =[99, 86, 87, 88, 100, 86,
    103, 87, 94, 78, 77, 85, 86]

plt.scatter(x, y, c ="blue")

# To show the plot
plt.show()
````

**<h2><center>SUBPLOTS</center></h2>**

Sublots means group of axes presents in a single figure. Matplotlib library has subplots function helps to plot differents types of plot in a single figure.

**SUBPLOTS()**

This function helps to present different axes in a single plot.

```
Syntax :

        matplotlib.pyplot.subplots(nrows=1,
                                   ncols=1,
                                   sharex=False,
                                   sharey=False
                                   )
        nrows, ncols : These parameter are the number of rows/columns
        of the subplot grid.
        sharex, sharey : These parameter controls sharing of properties
        among x (sharex) or y (sharey) axes.
```

````python
fig, axes = plt.subplots(nrows = 1,
                         ncols = 2)

axes[0].scatter(x, y, c ="blue")
axes[1].boxplot(data)
````

**set_title()** We can set title for each axis

**set-xlabel()** we can set x axis

**set-ylabel()** we can set y axis

````python
fig, axes = plt.subplots(nrows = 1,
                         ncols = 2)

axes[0].scatter(x, y, c ="blue")
axes[0].set_title("scatter plot example")
axes[0].set_xlabel("X - axis")
axes[0].set_ylabel("Y - axis")

axes[1].boxplot(data)
axes[1].set_title("bBox plot example")
axes[1].set_xlabel("X - axis")
axes[1].set_ylabel("Y - axis")
````

you can change figure size.

```
Syntax :
        plt.figure(figsize=(x,y))
```

````python
fig, axes = plt.subplots(nrows = 1,
                         ncols = 2,
                         figsize=(10, 5))

axes[0].scatter(x, y, c ="blue")
axes[0].set_title("scatter plot example")
axes[0].set_xlabel("X - axis")
axes[0].set_ylabel("Y - axis")

axes[1].boxplot(data)
axes[1].set_title("bBox plot example")
axes[1].set_xlabel("X - axis")
axes[1].set_ylabel("Y - axis")
````
