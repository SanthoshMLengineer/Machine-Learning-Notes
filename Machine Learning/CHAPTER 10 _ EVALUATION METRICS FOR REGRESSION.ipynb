**<h1><center>EVALUATION METRICS FOR REGRESSION</center></h1>**

Regression is a task of predicting continuous values depending on pattern observed while training. After creating the model for regression the next step is to find out how effective is the model based on some Evaluation metric using test datasets.

<U>Evaluation Metrics used : </U>

1. Mean Squared Error(MSE)

2. Root-Mean-Squared-Error(RMSE).

3. Mean-Absolute-Error(MAE).

4. R² or Coefficient of Determination.

**<u>Mean Squared Error</u>** :     

It is average of the squared difference between the output value and the value predicted by the regression model. Squaring is neccessary to remove negative signs. 

```
Mean Squared Error  = (1/n) * Σ(actual – predicted)2

n is number of records

actual is actual value

predicted is predicted value by the regression model
```

We can use Skearn Package to find the mean_squared_error , where y_true is actual value and y_pred is output from the model.

```
from sklearn.metrics import mean_squared_error

mean_squared_error(y_true, y_pred)
```

**<u>Root Mean Squared Error</u>**

It is just square root of mean squared error means square root of the averaged squared difference between the target value and the value predicted by the model. 

It can be done just square rooting mean squared error

```
from sklearn.metrics import mean_squared_error
import numpy as np

np.sqrt(mean_squared_error(y_true, y_pred))
```

**<u>Mean Absolute Error</u>**

Mean Absolute Error is average of absolute difference between actual value and predicted value. 

```
    Mean Absolute Error = (1/n) * ∑|Actual value – Predicted value|

    where,
        Actual value : Actual value for the observation
        predicted value : Calculated value for the observation from regression line
        n: Total number of observations
```

We can use Skearn Package to find the mean_absolute_error , where y_true is actual value and y_pred is output from the model.

```
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_true, y_pred)
```

**<u>R² or Coefficient of Determination</u>**

R Square measures how much of variability in dependent variable can be explained by the model. It is square of Correlation Coefficient(R) and that is why it is called R Square.
R Square is calculated by the sum of squared of prediction error divided by the total sum of square difference between actual value and the mean. R Square value is between 0 to 1 and bigger value indicates a better fit between prediction and actual value.

We can use Skearn Package to find the r2_score , where y_true is actual value and y_pred is output from the model.

```
from sklearn.metrics import r2_score

r2_score(y_true, y_pred)
```
