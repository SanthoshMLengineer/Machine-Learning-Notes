**<h1><center>LOGISTIC REGRESSION</center></h1>**

Logistic Regression is a supervised machine learning algorithms. It is used for classification problem.
Example : Predicting whether patient has cancer or not.

**Logistic Regression Assumptions:**
- Binary logistic regression requires the dependent variable to be binary.
- The independent variables should be independent of each other. That is, the model should have little or no multi-collinearity.
- The independent variables are linearly related to the log odds.
- Logistic regression requires quite large sample sizes.




The basis for this algorithm is sigmoid function and it is also called logistic function. Sigmoid function will give output in between of 0 and 1 for any real valued number. Logistic regression model takes a linear equation as input and uses logistic function and log odds to perform a binary classification task.

This function will return values 0 to 1, and we need to set threshould value to classify into classes. Normally we set threshold to 0.5, if output of logistic function more than 0.5 classified into `1` class and if value is less than 0.5 classified into `0` class.

**Types of Logistic Regression**

1. <u>Binary Logistic Regression</u> :
The categorical response has only two 2 possible outcomes. Example: Spam or Not
2. <u>Multinomial Logistic Regression</u> :
Three or more categories without ordering. Example: Predicting which food is preferred more (Veg, Non-Veg, Vegan)
3. <u>Ordinal Logistic Regression</u> :
Three or more categories with ordering. Example: Movie rating from 1 to 5
