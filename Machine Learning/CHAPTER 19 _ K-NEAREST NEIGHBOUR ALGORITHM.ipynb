**<h1><center>K-NEAREST NEIGHBOUR  ALGORITHM</center></h1>**

K-Nearest Neighbour is a supervised machine learning algorithm, which can be used for both classification and regression. It is also called lazy learning algorithm because it will not learn hypothesis function from training data, instead it memorize data. This algorithm doesn't assume anything about data this is non-parametric algorithm.

<u>Algorithm steps</u>  :  

1. Load the training data.
2. Initialize the K to choose the number of neighbours.
3. For each observation in test data
  - Calculate the distance between test data and each row of training data, add distance and row no of training data to the ordered collection.
  - Sort the ordered collection of distances from smallest to largest by the distances.
  - Select first K entries from sorted distance.
  - Get the labels of First K entries.
  - If Regression, return mean of labels.
  - If classification, return mode of labels.


**<u>Calculating Distance</u>**

**<u>Other popular distance measures</u> :**

1. Hamming Distance: Calculate the distance between binary vectors.

2. Manhattan Distance: Calculate the distance between real vectors using the sum of their absolute difference. Also called City Block Distance.

3. Minkowski Distance: Generalization of Euclidean and Manhattan distance.
  
**<u>Choosing Right value of K</u>**

- There is no structured method to find the best value for “K”. We need to tune this hyperparameter using tuning methods.
- Try and keep the value of k odd in order to avoid confusion between two classes of data.
- In general, practice, choosing the value of k is k = sqrt(N) where N stands for the number of samples in your training dataset.


**<u>Advantages of K-NN</u> :**
- The K-NN algorithm is very easy to implement.
- Lends itself very easily to parallel implementation.


**<u>Disadvantages of K-NN</u> :**

- Large storage requirements.
- Highly susceptible to the curse of dimensionality.
