**<h1><center>LINEAR REGRESSION</center></h1>**

Linear Regression is a supervised machine learning algorithm and it solves regression problem. It trains on labelled input data and based on observed pattern it produces output. It tries to find linear relationship between independent variable and dependent variable.

**<u>Assumptions of Linear Regression</u>**

1. Their should be linear relationship between dependent variables and independent variable.

2. Their should no Multicolinearity in independent variables means independent variable should not have correlation.

3. Homoscedasticity of errors means varience of residuals or errors is constant.Error value does not vary much as the value of the predictor variable changes.

4. Errors should be normally distributed.


**<u>Hypothesis of Linear Regression</u>**

```
y = B0 + B1x

y is dependent variable or output variable

x is independent variable 

B0 is intercept of line

B1 is coefficient of X
```



**<u>Their are two types of linear regression</u>**

1. Simple Linear Regression

2. Mulitiple Linear Regression

<u>1. Simple Linear Regression</u>

If their is only one independent variable is used to predict dependent variable then such linear regression algorithm is called simple linear regression.

<u>2. Multiple Linear Regression</u>

If more than one independent variable is used to predict dependent variable then such linear regression algorithm is called Multiple Linear Regression.

**<u><h3>How to get best fit line</h3></u>**

Our main aim is to find best fit line means the error between predicted value and actual value is low. We will use cost function in order get best coefficients and intercept values and these values will have minimum error between actual value and predicted value.

**Mean Squared Error** function is used as cost function.

**<u>Gradient Descent</u>**

Regression model use Gradient Descent Algorithm to update coefficients of line by cost function. It selects random values of coefficients and it updates the value of coefficients to reach minimum value of cost function.
