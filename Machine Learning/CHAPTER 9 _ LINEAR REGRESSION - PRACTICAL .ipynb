Sklearn is machine learning library for python language. It has various classification, regression algorithms and various datasets are available. Here we will load load_boston dataset and train linear regression using sklearn library.


```python
from sklearn.datasets import load_boston

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

import pandas as pd

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

regre_data = load_boston()

print(regre_data.DESCR)
```

```python
# Training Data
data = pd.DataFrame(regre_data.data,columns = regre_data.feature_names)

# Testing Data
data['target'] = regre_data.target

data.head()
```

```python
corr = data.corr()
```

```python
print (corr['target'].sort_values(ascending=False)[:5], '\n') #top 15 values
print ('----------------------')
print (corr['target'].sort_values(ascending=False)[-5:]) #last 5 values`
```

```python
X = data.loc[:,['RM','ZN','B','DIS','NOX','TAX','INDUS','PTRATIO','LSTAT']]
y = data['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)

#Creating the Model
model = LinearRegression(normalize = True)

# Training the model
model.fit(X,y)
```

```python
print("Coef:", model.coef_)
print("Intercept:", model.intercept_)
```

Here we can see coefficients and intercept of regression line. Number of coefficients will be equal to number of independent variable.
