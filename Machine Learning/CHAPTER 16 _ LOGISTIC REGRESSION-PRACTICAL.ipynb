```python
import pandas as pd
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
data = pd.read_csv("//content//drive//My Drive//Personal Projects//Diabetics \
Prediction//Diabetes Prediction//diabetes.csv")

data.shape
```

```python
data.head()
```

Outcome column is target variable where 0 indicates non-diabetic and 1 indicates diabetic.


```python
sns.countplot(data['Outcome'])
```

```python
data['Outcome'].value_counts()/len(data)
```

Their is slightly imbalanced in data. Where `0` class has majority of where more data belongs to this class and `1` minority where less data belongs to this class.

```python
X = data.drop('Outcome',axis = 1)
y = data['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=1, 
                                                    stratify=y)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

log_cross_val_log = cross_val_score(log_reg,X_train, y_train,
                                    cv=10,scoring="f1_macro")
```

```python
print("Accuracy of logistic regression is {}".format(log_reg.score(X_test,
                                                                   y_test)))

print("Cross Validation score is ",log_cross_val_log)
```

This dataset is imbalanced, we need to treat them so that it will effect on accuracy.
