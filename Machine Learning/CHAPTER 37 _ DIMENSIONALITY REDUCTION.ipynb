**<h1><center>DIMENSIONALITY REDUCTION</center></h1>**

Dimensionality Reduction is a process of reducing number of features in the dataset. If an dataset has more than hundred of features, then number of samples also increases with all combination this makes model to overfit.


<u>The Curse of Dimensionality</u>

<center><img src="https://miro.medium.com/max/510/1*vah8IolNqlxNHq9ysVzYkw.png" width = 400px></center>

As the number of features increases, the model becomes more complex. The more the number of features, the more the chances of overfitting. 

Dimensionality reduction helps to avoid overfit. The fewer features our training data has, the lesser assumptions our model makes and the simpler it will be.

Feature Reduction Technique : 

1. Principle Component Analysis

2. Linear Discremanant Analysis

<u>**1. Principle Component Analysis**</u> : 

Principal Components Analysis is an unsupervised learning used for reducing dimensionality. This  technique helps to find out important variables. 



<u>Principle component Analysis using Sklearn</u>

```python
# Importing the required Package
from sklearn import datasets
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Load data
iris = datasets.load_iris()

# Description About data set
print(iris.DESCR)

# After seeing below we need to build a model for classifying the classes of iris, so it is classification Problem.
```

```python
# Training Data
train = pd.DataFrame(iris.data,columns = iris.feature_names)

# Testing Data
target = pd.DataFrame(iris.target,columns = ['species'])

print(train.head())
print(target.head())
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,
                                                 target,
                                                 test_size=0.3)

model = RandomForestClassifier()
model.fit(X_train,y_train)

y_predict = model.predict(X_test)

print("accuracy without pca analysis",
      accuracy_score(y_test,y_predict))
```

<u>Using PCA</u>

```python
# apply PCA
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(train)
X_pca = pd.DataFrame(X_pca,columns=['pca1','pca2','pca3','pca4'])
X_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
X_pca_select = X_pca.iloc[:,[0,1]]

X_pca_select.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_pca_select,
                                                 target,
                                                 test_size=0.3)
model = RandomForestClassifier(n_estimators=12)

model.fit(X_train,y_train)
y_predict = model.predict(X_test)
accuracy_score(y_test,y_predict)
```

**<u>2. Linear Discremanant Analysis</u>**

It is used for Supervised classification Problems. LDA is a supervised data compression technique which is aimed increasing class distinction techniques.


```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda = LDA(n_components=2)  
X_train,X_test,y_train,y_test = train_test_split(train,
                                                 target,
                                                 test_size=0.3)

y_train.value_counts()
```

```python
X_train = lda.fit_transform(X_train, y_train)  
X_test = lda.transform(X_test)  
```

```python
model.fit(X_train, y_train)  
y_pred = model.predict(X_test) 

accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,
                            y_pred))
```
