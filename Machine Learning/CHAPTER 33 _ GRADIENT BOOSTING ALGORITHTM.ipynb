**<h1><center>GRADIENT BOOSTING ALGORITHM</center></h1>**

Gradient Boosting algorithm is one of the ensemble methods where group of weak learners are combined to give better results. This algorithm is used for both classification and regression.

**<h3><u>GRADIENT BOOSTING ALGORITHM FOR REGRESSION</u></h3>**

<u>Steps</u>

1. Initially tree has only one node that is leaf node. And value of that node is average of target variable.

2. Now we will calculate residual(`Residual_from_first_tree` for first tree residual ) for each data point. Now we will train decision tree to predict these `Residual_from_first_tree` values. (In Gradient Boosting Algorithm can have little deeper trees with 8 to 32 terminal nodes)

3. total_prediction_upto_one = Prediction of initial tree + Learning rate * (Prediction of second tree). Now calculate the residual by taking Target variable and total_prediction_upto_one. Now we will consider this residual (Residual_from_second_tree).

**Note** : Learning rate is multiplied to every tree predictions except initial tree this helps to move in a right direction with a small step.

4. Now we will create another decision tree to predict Residual-2 and repeat stpes from 2.


**<h3><u>GRADIENT BOOSTING ALGORITHM FOR CLASSIFICATION</u></h3>**

<u>Steps</u> :

1. <u>First Calculate log of odds :</u>

  Odds is the ration of something happening to something not happening. Let us assume one Example
  ```
  -------  ------  --------  --------    ---------
  Pclass    Age     Fare      Sex         Survived
  -------  ------  --------  --------    ---------
   3         22     7.25      male          0
  -------  ------  --------  --------    ---------
   1         38     71.2833   female        1
  -------  ------  --------  --------    ---------
   2         26     7.925     female        1
  -------  ------  --------  --------    ---------
   1         35     53.1001   female        1
  -------  ------  --------  --------    ---------
   3         8      21.07     male          0
  -------  ------  --------  --------    ---------
   3         7      11.133    female        1
  -------  ------  --------  --------    ----------
  ```
  Here we assume something happening is `1` class and `0` class.
  
  Now calculate log of odds = something happening / Something not Happening
    
    = log base`e`(4/6 / 2/6)

    = log base`e`(4/2)

    = log base`e`(2)

    = 0.7

2. <u>Calculate initial prediction (probability) using log of odds</u> :    

  The easiest way to use the log(odds) for classification is to convert it to a probability using logistic function.

  P = e^log(odds) / 1 + e^log(odds)
    
    = e ^ 0.7 / 1 + e ^ 0.7

    = 2.01375270747 / 1 + 2.01375270747

    = 2.01375270747 / 3.01375270747

    = 0.668 nearly 0.7

  If Probability is greater than 0.5 then we classify into class `1` and if it is less than 0.5 then we classify into `0`.

3. <u>Calculate Psudo Residual : </u>

  Here we will calculate psudo residual, the difference between observed and probability value.
  ```
  -------  ------  --------  --------    ---------   ----------------
  Pclass    Age     Fare      Sex         Survived    Residual
  -------  ------  --------  --------    ---------   -----------------
   3         22     7.25      male          0         0 - 0.7 = -0.7
  -------  ------  --------  --------    ---------   -----------------
   1         38     71.2833   female        1         1 - 0.7 = 0.3
  -------  ------  --------  --------    ---------   -----------------
   2         26     7.925     female        1         1 - 0.7 = 0.3
  -------  ------  --------  --------    ---------   -----------------
   1         35     53.1001   female        1         1 - 0.7 = 0.3
  -------  ------  --------  --------    ---------   -----------------
   3         8      21.07     male          0         0 - 0.7 = -0.7
  -------  ------  --------  --------    ---------   -----------------
   3         7      11.133    female        1         1 - 0.7 = 0.3
  -------  ------  --------  --------    ----------  -----------------
  ```
