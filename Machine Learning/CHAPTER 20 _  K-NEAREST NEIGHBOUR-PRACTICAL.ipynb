**<u><h1>IRIS FLOWER CLASSIFICATION USING KNN</h1></u>**



```python
# Importing the required Package
import warnings
warnings.filterwarnings('ignore')

from sklearn import datasets

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold

import pandas as pd
```

```python
KNeighborsClassifier()
```

KNeighborsClassifier having many the parameters, the most important parameters is n_neighbors. We will tune n_neighbors parameter using the hyperparameter techniques.

```python
# Load data
iris = datasets.load_iris()
```

```python
# Description About data set
print(iris.DESCR)
```

```python
# Training Data
train = pd.DataFrame(iris.data,columns = iris.feature_names)

# Testing Data
target = pd.DataFrame(iris.target,columns = ['species'])

print(train.head())
print(target.head())
```

```python
X_train,X_test,y_train,y_test=train_test_split(train,
                                               target,
                                               test_size=0.3)

model=KNeighborsClassifier(n_neighbors=40)

model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)

print("Accuracy of KNN model for classifying iris species",
      accuracy_score(y_test,y_predict),"\n")

print(classification_report(y_test,y_predict))
```

**K-Fold Cross Validation**

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)

k_fold_score = []
for train_index, test_index in kf.split(train,target):
    
    # print(X.iloc[list(train_index),:])
    
    X_train, X_test = train.iloc[list(train_index),:], train.iloc[list(test_index),:]
    y_train, y_test = target.iloc[list(train_index),:], target.iloc[list(test_index),:]
    
    model = KNeighborsClassifier(n_neighbors=11)
    model.fit(X_train,y_train)
    y_predict = model.predict(X_test)
    k_fold_score.append(accuracy_score(y_test,y_predict))
   
```

```python
import statistics

print("Minimum accuracy we get is {}".format(min(k_fold_score)))
print("Maximun accuracy we get is {}".format(max(k_fold_score)))
print("We can get average accuracy is {}".format(
    statistics.mean(k_fold_score)))
```

We need to tune k value as their is fixed value to which we will get good results we need to do trial and error method for different values.
