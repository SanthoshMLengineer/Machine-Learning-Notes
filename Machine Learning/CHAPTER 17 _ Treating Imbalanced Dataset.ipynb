**<h1><center>IMBALANCED DATASET</center></h1>**

Unequal distribution of classes within dataset is considered to be Imbalanced. If we train this type of dataset then model will be baised. 

<u>Techniques to solve imbalance problem</u>

1. Undersampling

2. Oversampling

**<u>1. Undersampling</u>**

Undersampling is the process where you randomly delete some of the observations from the majority class in order to match the numbers with the minority class.

The disadvantage of this is loosing information as we are deleting the observation. This may lead to underfitting because model will not generalise from data with less information.

Example : Let us discuss credit card fraud detection.Their are 492 observation belongs to the fraud class that is minority class. We will get only 492 sample from the majority class that is non - fraud class. And then we cancat this to make single data frame.

```
            # Shuffle the Dataset.
            shuffled_df = credit_df.sample(frac=1,random_state=4)

            # Put all the fraud class in a separate dataset.
            fraud_df = shuffled_df.loc[shuffled_df['Class'] == 1]

            #Randomly select 492 observations from the non-fraud (majority class)
            non_fraud_df = shuffled_df.loc[shuffled_df['Class'] == 0].sample(n=492,random_state=42)

            # Concatenate both dataframes again
            normalized_df = pd.concat([fraud_df, non_fraud_df])
```

**<u>2. Oversampling</u>**

Oversampling is generating the synthetic data of minority class in order to match majority class.

SMOTE(Syntetic Manority Oversampling Technique) is a technique used for Oversampling.

Example : Let us discuss credit card fraud detection.Their are 492 observation belongs to the fraud class that is minority class. And Non - fraud class is having 5000 observation. Now we will increase minority class to match majority class.
```         
from imblearn.over_sampling import SMOTE

# Resample the minority class. You can change the strategy to 'auto' if you are not sure.
sm = SMOTE(sampling_strategy='minority', random_state=7)

# Fit the model to generate the data.
# X is feature and y is target
oversampled_trainX, oversampled_trainY = sm.fit_resample(X, y)

oversampled_train = pd.concat([pd.DataFrame(oversampled_trainY), pd.DataFrame(oversampled_trainX)], axis=1)
```
