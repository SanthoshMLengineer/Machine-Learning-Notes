**<center><h1>FEATURE SCALING</h1></center>**

Feature Scaling is a process of shifting the values of a feature to a fixed range. Feature scaling is an important process while we dealing with distance based algorithms and Gradient descent based algorithm.

- Distance algorithms like KNN, K-means, and SVM are most affected by the range of features. This is because behind the scenes they are using distances between data points to determine their similarity.

- Tree-based algorithms will not effected by different ranges of features. A decision tree is only splitting a node based on a single feature and this feature is not influenced by other. 

Their are two types of Scaling

- Normalization or Minmax Scaling

- Standaration

**Normalization or Min-Max Scaling**

In this technique values are rescaled in the range of 0 to 1. This can be used when feature is not normally distributed.

  <center><img src="https://i0.wp.com/analyticsindiamag.com/wp-content/uploads/2021/08/image-288.png?resize=1024%2C253&ssl=1" Width = 500px></center>

Here min(x) and max(x) is the minimum and maximum value of feature respectively. 

When the value of X is the minimum value in the column, the numerator will be 0, and hence X’ is 0. On the other hand, when the value of X is the maximum value in the column, the numerator is equal to the denominator and thus the value of X’ is 1. If the value of X is between the minimum and the maximum value, then the value of X’ is between 0 and 1.

```python
import numpy as np
import seaborn as sns


df=np.array([[2, 3, 7, 30],
             [9, 4, 6, 1],
             [8, 15, 2, 40],
             [20, 10, 2, 6]]
            )

# It is not normally distributed 

# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df)
scaled_features = scaler.transform(df)
print(scaled_features)

```

**Standardation**

This scaling technique can be used when feature is normally distributed. In this technique values in the feature are replaced with z-score. Z-score is also known as standard score gives us how many standard deviations an element is from the mean. Hence, Z-Score is measured in terms of standard deviation from the mean.

<center><img src="https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/Stand_eq.gif" width = 200px></center>

mu is the mean of the feature values, sigma is the standard deviation of the feature.



```python
import numpy as np
import seaborn as sns


df=np.array([[2, 3, 7, 30],
             [9, 4, 6, 1],
             [8, 15, 2, 40],
             [20, 10, 2, 6]]
            )

from sklearn.preprocessing import StandardScaler

scale = StandardScaler().fit(df)
    
# transform the training data column
X_train_stand = scale.transform(df)

print(X_train_stand)
sns.distplot(X_train_stand)
```
