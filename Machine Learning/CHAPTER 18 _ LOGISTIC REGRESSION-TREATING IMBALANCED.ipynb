```python
import pandas as pd
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
data = pd.read_csv("//content//drive//My Drive//SELLING COURSE AND PROJECTS//\
REAL TIME PROJECTS//Diabetes Prediction//Data Files//diabetes.csv")

data.shape
```

```python
data.head()
```

Outcome column is target variable where 0 indicates non-diabetic and 1 indicates diabetic.


```python
data['Outcome'].value_counts()/len(data)
```

Their is slightly imbalanced in data. Where `0` class has majority of where more data belongs to this class and `1` minority where less data belongs to this class.

```python
X = data.drop('Outcome',axis = 1)
y = data['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.20,
                                                    random_state=1,
                                                    stratify=y)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

log_cross_val_log = cross_val_score(log_reg,X_train, y_train,
                                    cv=10,scoring="f1_macro")

```

```python
from sklearn.metrics import roc_auc_score

pred = log_reg.predict(X_test)

auc_score = roc_auc_score(y_test, pred)
```

```python
print("AUC SCORE of logistic regression is {}".format(auc_score))

print("Cross Validation score is ",log_cross_val_log)
```

This dataset is imbalanced, we need to treat them so that it will effect on accuracy.

**Treating Imbalanced Dataset**

```python
from imblearn.over_sampling import SMOTE

# Resample the minority class. You can change the strategy to 'auto' if you are not sure.
sm = SMOTE(sampling_strategy='minority', random_state=7)

# Fit the model to generate the data.
# X is feature and y is target
oversampled_trainX, oversampled_trainY = sm.fit_resample(X_train, y_train)

```

```python
log_reg = LogisticRegression()
log_reg.fit(oversampled_trainX, oversampled_trainY)

log_cross_val_log = cross_val_score(log_reg,oversampled_trainX,
                                    oversampled_trainY,
                                    cv=10,scoring="f1_macro")
```

```python
pred = log_reg.predict(X_test)

auc_score = roc_auc_score(y_test, pred)

print("AUC SCORE of logistic regression is {}".format(auc_score))

print("Cross Validation score is ",log_cross_val_log)
```

Their is no much impact on model after treating imbalanced dataset.
